```python
import requests # 웹 페이지 소스를 얻기 위한 패키지(기본 내장 패키지이다.)
from bs4 import BeautifulSoup # 웹 페이지 소스를 얻기 위한 패키지, 더 간단히 얻을 수 있다는 장점이 있다고 한다.
                              # (!pip install beautifulsoup4 으로 다운받을 수 있다.)
import pandas as pd # 데이터를 처리하기 위한 가장 기본적인 패키지
import time # 사이트를 불러올 때, 작업 지연시간을 지정해주기 위한 패키지이다. (사이트가 늦게 켜지면 에러가 발생하기 때문)
import urllib.request # 이미지 파일을 저장하기 위한 패키지, 웹사이트의 이미지 주소가 있으면 다운 받을 수 있다.
```

```python
# 카테고리에 대한 주소를 반복문을 통해 얻는 것 보다 카테고리 주소를 직접 저장하는 것이 간편하여 실행

base_url = 'http://www.superstari.co.kr'
outer_ = 'http://www.superstari.co.kr/shop/shopbrand.html?xcode=003&type=Y&gf_ref=Yz04OUZsbXU='
top_ = 'http://www.superstari.co.kr/shop/shopbrand.html?xcode=156&type=Y&gf_ref=Yz1qM1U5cGM='
bottom_ = 'http://www.superstari.co.kr/shop/shopbrand.html?xcode=134&type=Y&gf_ref=Yz1teWJ2M0I='
shirt_ = 'http://www.superstari.co.kr/shop/shopbrand.html?xcode=142&type=Y&gf_ref=Yz1sS1JRVmY='
training_ = 'http://www.superstari.co.kr/shop/shopbrand.html?xcode=119&type=Y&gf_ref=Yz1aUFlsTTg='

category_list_url = [outer_, top_, bottom_, shirt_, training_]
```

```python
start_url = top_
start_url_name = 'top_'
page_url = start_url + '&page=' + '{}'
    
page_1 = requests.get(start_url)
page_2 = BeautifulSoup(page_1.content, 'html.parser')
page_3 = page_2.find('div', {'class' : 'item-page'}) # 상품 페이지 클래스 획득
page_4 = page_3.find_all('a')
page_5 = len(page_4) 
```

```python
list = pd.DataFrame({'name' : [], 'img' : []}) # 데이터를 추가하기 위해 list라는 데이터프레임을 미리 저장

start_url = top_
start_url_name = 'top_'
page_url = start_url + '&page=' + '{}'
    
page_1 = requests.get(start_url)
page_2 = BeautifulSoup(page_1.content, 'html.parser')
page_3 = page_2.find('div', {'class' : 'item-page'}) # 상품 페이지 클래스 획득
page_4 = page_3.find_all('a')
page_5 = len(page_4) 
# 페이지 수를 각 카테고리 별로 일일히 다르게 주기보다 페이지 수의 소스를 가지고와 자동으로 추출하기 위해 진행하였다.
    
for page in range(1,page_5) :
    res = requests.get(page_url.format(page))
    html = BeautifulSoup(res.content, 'html.parser')
    table = html.find_all('td', {'valign' : 'top'}) # 각각의 상품에 접근하기 위해 클래스획득
        
    products = []
    for tb in range(len(table)) :
        prod = table[tb].find('a')['href'] # 카테고리의 href를 가지고 온것 처럼 한 카테고리의 한 상품에 대한 href를 가지고 오는 것
        if prod not in products :
            products.append(prod)
        
    url_list = []
    for pr in range(len(products)) :
        url = base_url + products[pr] # 이를 base_url과 합쳐서 각각의 상품에 접근할 수 있는 url을 획득한다.
        url_list.append(url)
            
    for i in range(len(url_list)) :
        res = requests.get(url_list[i])
        html = BeautifulSoup(res.content, 'html.parser')
        table = html.find_all(True, {'class' : ['detail_pic_1_2', 'detail_pic_2_2']})  # 각 상품의 이미지를 추출하기 위한 클래스획득
        
        url = url_list[i]
        
        names = []
        for j in range(len(table)) :
            alt = table[j].find_all('em')
            for k in range(len(alt)) :
                name = 'prod_' + alt[k].text + '_' + str(page) + '_'+ str(i) 
                # 한 상품에 색상별로 이미지가 여러가지 있으므로 이를 각각 저장하기 위해 이름이 따로 저장
                if name :
                    names.append(name)
                
        imgs = []
        for tb in range(len(table)) :
            src = table[tb].find_all('img')
            for sr in range(len(src)) :
                img = src[sr].get('src') # 이미지의 주소에 대한 정보는 img 태그 내의 src 속성에 있다. 이를 가지고 오는 것.
                if img :
                    imgs.append(img)
                    
        if len(names) == len(imgs) :
            df = pd.DataFrame({'name' : names, 'img' : imgs, 'url' : url}) 
            # 상품명/색상명 과 이미지주소, 해당 상품의 url(사이즈표를 가지고 오기 위한) 을 데이터 프레임 형식으로 저장
            
        for id in df.index :
            if '뒷면' in df['name'][id] : # 뒷면이라는 이미지도 있어서 뒷면이라는 이미지 이름을 가진 것을 빼고 list에 저장한다.
                df = df.drop(id, axis = 0)   
            
        list = list.append(df)
```

```python
list.index = range(len(list)) # 데이터를 append한 것이기 때문에 인덱스가 엉망이므로, 재설정을 해줌.
```

```python
try :
    for i in range(len(list)) : # 옷 이미지에 대한 이미지 파일을 저장하는 것, 반복문을 통해 하나씩 저장해주었다.
        url = list['img'].iloc[i]
        name = start_url_name + str(i) + list['name'].iloc[i] + '_img' + '.jpg' # 티100상품명_img.jpg 형식으로 저장
        urllib.request.urlretrieve(url, name)
        
except urllib2.HTTPError or error :
    pass
```

```python
from selenium.webdriver import Chrome 
# 사이즈표 형식이 텍스트형식이고, html 안에 html이 존재하는 상품페이지가 있어서 beautifulsoup4로 데이터 수집이 불가능하다.
# 따라서, selenium을 사용해 데이터를 수집한다. 
# (!pip install selenium 로 패키지를 다운받는다.)
# 셀레늄 브라우저별 웹드라이버 링크는 셀레늄의 다운로드 페이지에서 다운받는다. 
# 크롬 웹드라이버 = ('https://sites.google.com/a/chromium.org/chromedriver/downloads')

prds = []
index = []
for i in range(len(list)) :
    prd = list['url'][i]
    ind = list.index[i]
    if prd not in prds : # 중복되는 상품이 없도록 url을 수집한다.
        prds.append(prd) 
        index.append(ind) # 해당 상품에 대한 인덱스 정보가 있어야 이후에 옷 이미지와 매칭시킬 수 있으므로 인덱스 정보까지 포함한다.

browser = Chrome() # 크롬 브라우저를 연다.

for web in range(len(prds)) : # 저장된 url에서 사이즈표를 추출하기 위해 반복문을 사용한다.
    num = index[web]
    browser.get(prds[web]) # prds에 저장된 url을 순서대로 접속한다.
    time.sleep(5) # 작업 지연시간을 지정해주기 위한 것.(없으면 브라우저를 열기도 전에 class를 탐색해 에러가 발생할 것이다.)
    
    dsize = browser.find_elements_by_css_selector('div.dsize_dsize') # 사이즈표가 하나의 html안에 있는 것, 바로 접근가능
    iframe = browser.find_elements_by_css_selector('#sf_chart_iframe_id') # 사이즈표가 또다른 html안에 있는 것, 우회해서 접근해야한다.
    
    if dsize : # 사이즈표가 하나의 html안에 있는 경우
        for i in range(len(dsize)) :
            tr = dsize[i].find_elements_by_tag_name('tr') # 사이즈표가 있는 table 태그의 tr속성을 찾는다. -> 사이즈표의 각 행을 의미함.
            
            tds = []
            for i in tr :
                td = i.find_elements_by_tag_name('td') # 사이즈표의 각각의 텍스트 데이터를 추출하는 것을 의미한다.
                tds.append(td)
                
            tx = []
            txt = []
            
            for i in range(len(tr)) :
                for j in range(len(tds[i])) :
                    tx = tds[i][j].text # td에 저장되어 있는 텍스트 데이터를 가지고 와서 txt에 저장한다.
                    txt.append(tx)
                    
            start_pos = 0

            end_pos = len(txt)

            div = len(tr[0].find_elements_by_tag_name('td')) # txt가 하나의 리스트로 저장이 되기 때문에 
            # 이를 데이터프레임화해주기 위해 리스트를 잘라서 다시 저장하여야 한다. 따라서, 사이즈표의 열개수를 지정해주는 것

            data = []

            for idx in range(start_pos, end_pos + div, div): # 리스트를 잘라서 다시 저장하는 것
                output = txt[start_pos:start_pos + div]
                if output :
                    data.append(output)
                start_pos = start_pos + div
            
            df_dsize = pd.DataFrame(data[1:], columns = data[0]) # Multi list를 DataFrame으로 바꿔주는 것
            
            name = 'prod_' + ' top_' + str(num) +'_'+ str(web) + '_size' + '.csv' 
            # ex) prod_top_index_prds_size.csv 형식으로 저장한다. num 은 각각의 상품에 대한 인덱스를 의미, web은 각각의 다른 상품을 의미
            df_dsize.to_csv(name, mode='w', encoding='ANSI',index=False) # 데이터프레임을 csv형식으로 저장
    
    elif iframe : # 사이즈표가 또다른 html안에 있는 경우
        
        browser.switch_to_frame(browser.find_element_by_css_selector('#sf_chart_iframe_id')) 
        # 또 다른 html에 접근하기 위해 해당 html에 대한 정보를 가지고 있는 iframe id 를 획득
        # .switch_to_frame 은 iframe 내부에 접근하는 것.
        source = browser.page_source # iframe 내부로 접근하면 텍스트 형식이기 때문에, 이를 html로 파싱해주어야 한다.
    
        html = BeautifulSoup(source, 'html.parser')
    
        tbody = html.find('tbody') # 사이즈표가 있는 tbody 태그를 추출
    
        tr = tbody.find_all('tr') # 사이즈표의 행을 의미하는 tr을 추출
    
        div = len(tr[0].find_all('th')) # txt가 하나의 리스트로 저장이 되기 때문에 
        # 이를 데이터프레임화해주기 위해 리스트를 잘라서 다시 저장하여야 한다. 따라서, 사이즈표의 열개수를 지정해주는 것
    
        txt = []
        for i in range(len(tr)) :
            for j in range(len(tr[i])) :
                tx = tr[i].find_all(['th', 'td'])[j].text # td에 저장되어 있는 텍스트 데이터를 가지고 와서 txt에 저장한다.
                txt.append(tx)
            
        start_pos = 0

        end_pos = len(txt)

        data = []

        for idx in range(start_pos, end_pos + div, div): # 리스트를 잘라서 다시 저장하는 것
            output = txt[start_pos:start_pos + div]
            if output :
                data.append(output)
            start_pos = start_pos + div
            
        df_iframe = pd.DataFrame(data[1:], columns = data[0]) # Multi list를 DataFrame으로 바꿔주는 것
    
        name = 'prod_' + ' top_' + str(num) +'_'+ str(web) + '_size' + '.csv' 
        # ex) prod_top_index_prds_size.csv 형식으로 저장한다. num 은 각각의 상품에 대한 인덱스를 의미, web은 각각의 다른 상품을 의미
        df_iframe.to_csv(name, mode='w', encoding='ANSI',index=False) # 데이터프레임을 csv형식으로 저장
        
    else :
        pass
```

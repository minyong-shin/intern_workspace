### 텍스트 전처리

```python
#kkma 모듈을 사용할 것이기 때문에 이모지 형식을 다 제거해준다
def apply_kko_regex(msg_list):
    emoji_pattern = re.compile("["u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               "]+", flags=re.UNICODE)

    kko_parse_result = []
    
    for i in comment_list:
        tokens = re.sub(emoji_pattern,"",i)
        kko_parse_result.append(tokens)
    
    kko_parse_result = pd.DataFrame(kko_parse_result, columns=["comment"])
    
    return kko_parse_result
        
```

```python
tm0 = apply_kko_regex(comment_list)
```

### 형태소분석기를 통한 댓글 명사 추출

```python
#한글 형태소 분석기인 konlpy사용 
#이때 konlpy에는 hannanum, kommoran, kkma, twitter이 있는데 해당 분석에서는 kkma사용
from konlpy.tag import Kkma
#apply lambda사용
```

```python
#필요없는 문자열 제외
#re.compile 패턴객체의method를 부여하는 과정으로 4가지의 method를 제공
#match, search, findall,finditer
def get_noun(comment_txt):
    kkma = Kkma()
    nouns = []
    pattern = re.compile('[ㄱ-ㅎㅏ-ㅣ]+')
    comment_txt = re.sub(pattern,"",comment_txt).strip() #strip는 문자열로 반환하는 것
    
    #형태소 분석 시작
    if len(comment_txt) >0:
        pos = kkma.pos(comment_txt)
        #pos가 두개의 열로 나뉘기 때문에 반복문설정시 i,j로 설정
        for i,j in pos:
            #고유명사 OR 보통명사 -> type가 명사인것만 추출한 것임
            if j == "NNG" or j == "NNP":
                nouns.append(i)
        return nouns
```

```python
tm0["pc_text"] = tm0["comment"].apply(lambda x: get_noun(x))
```

### 추출한 명사를 이용한 wordcloud

```python
import pandas as pd
import ast
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
from wordcloud import WordCloud
```

```python
#단어들의 빈도수 측정
def draw_wordcloud(tm0):
    #ast.literal_eval = list로 되어 있는 열을 row단위로 분리
    #각 행마다 하나의 리스트 안에 여러개의 토큰이 존재하는데 리스트 안의 토큰을 풀어서 새로운 행을
    #구성하는 작업 - 근데 현재 오류가 심함..
    tokens = pd.DataFrame(tm0['pc_text'].apply(lambda x: ast.literal_eval(x)).tolist())
    
    tokens["comment"] = tm0["comment"]
    
    #인덱스 정리
    tokens = tokens.set_index(['comment','pc_text'])
    tokens = tokens.T.unstack().dropna().reset_index() #결측치처리 unstack??
#해당 함수를 만들지 않고 분해 완료
```

```python
#단어들만뽑아서 사용할 수 있지만 데이터프레임 형태로 뽑아보자
token_info = pd.DataFrame({'comment':[],
                          'token':[]})
for i in range(len(tm0)):
    comment = tm0['comment'].iloc[i]
    for j in range(len(tm0['pc_text'].iloc[i])): #각 댓글별 단어의 갯수
        token = tm0['pc_text'].iloc[i][j]
        insert_data =  pd.DataFrame({'comment':[comment],
                                     'token':[token]})
        token_info = token_info.append(insert_data)
```

# 벡터화된 데이터를 이용한 클러스터링 
### DBSCAN 알고리즘을 이용한 군집화 

dbscan은 밀도 기반의 클러스터링 기법임 
- 이웃하는 점들의 간격이 사용자가 정한 거리 이내에 있는 경우 산출
- 군집에 포함되는 점들이 최소 클러스터원소수(min_samples)보다 많은 경우만 산출

이러한 방법으로 군집화를 하게 된다면 클러스터의 갯수를 지정할 필요가 없으며 원형이 아닌 클러스터 형태로도 추출이 가능하다는 점이다.  
다른 알고리즘은 클러스터의 중심점을 이동하면서 클러스터 중심점 내부에 몇개의 점이 들어 있는 가에 집중하기 때문에 필히 원형 모양이 된다(kmeans)  
dbscan은 각각 개별 포인트 간의 거리를 측정하면서 **non-core-point**에 의해서 클러스터의 경계가 형성되기 때문에 원형이 아닌 모양(기하학적 모양)의 군집화가 가능한 것임  
BUT, if you don't understanding about data, eps , min_points to fix very hard
다른 알고리즘은 단순히 10개의 분류를 만들어 달라고 지시하면 10개를 만들어주지만 dbscan은 적어도 포인트 간 이루는 최대 거리가 얼마인지 정도는 파악해야 유효한 클러스터를 만들 수 있음

적절한 eps와 min_samples를 찾기 위해서 전처리가 꼭 필요

### 보는 것과 같이 적합한 epsilon과 min_samples를 찾는 것이 dbscan의 핵심이다  
### dbscan이 어떠한 방식으로 형성되고 어떻게 진행 되는지는 이해하고 있지만 최적의 epsilon과 min_sample를 찾는 것은 연습이 필요함 밑과 위의 코드는 주성분분석과 nmf성분으로 나눠서 따로따로 분석을 진행하고 최적의 값을 찾은 것이다 해당 코드들은 꼭 이해할 것!

```python
for min_sample in [3, 5, 7, 9]:

    for eps in [0.2, 0.4, 0.9, 1, 3, 5, 7, 9, 11, 13, 15, 17]:

        dbscan = DBSCAN(n_jobs=-1, min_samples=min_sample, eps=eps)# -1을 통해서 노이즈 제거

        labels_pca = dbscan.fit_predict(x_pca) # pca를 이용한 성분을 이용한 dbscan을 예측

        labels_nmf = dbscan.fit_predict(x_nmf) # nmf를 이용한 성분을 이용한 dbscan으로 예측

        print('︎=== min_sample:{}, eps:{} ==='.format(min_sample, eps))

        print('')

        print('pca ==> cluster 수: {}'.format(len(np.unique(labels_pca))))

        print('pca ==> cluster 크기: {}\n'.format(np.bincount(labels_pca+1)))



        print('nmf ==> cluster 수: {}'.format(len(np.unique(labels_nmf))))

        print('nmf ==> cluster 크기: {}'.format(np.bincount(labels_nmf+1)))

        print('---------------------------------------')
```

```python
# 가장 많은 클러스터가 생긴 eps와 n_sample 츨력

dbscan = DBSCAN(min_samples=3, eps=7, n_jobs=-1) # n_jobs = 코어의 갯수

labels_pca = dbscan.fit_predict(x_pca)



dbscan = DBSCAN(min_samples=3, eps=0.9, n_jobs=-1)

labels_nmf = dbscan.fit_predict(x_nmf)



# pca 주성분, DBSCAN 을 이용한 시각화

for cluster in range(labels_pca.max() + 1):

    idx = labels_pca == cluster 

    n_images = np.sum(idx) # idx중 true인 것의 갯수

    fig, axes = plt.subplots(1,n_images, figsize=(n_images*1.5, 4), # figsize=(a,b) axb 그림사이즈

                            subplot_kw={'xticks':(), 'yticks':()})

    

    for image, label, ax in zip(x_people[idx], y_people[idx], axes.ravel()):

        ax.imshow(image.reshape(image_size))

        ax.set_title(people.target_names[label].split()[-1])



plt.gray()

plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
model.wv.syn0.shape #w2v size
```

```python
word_vectors = model.wv.syn0 # model.wv.vectors와 같은 값을 가짐
```

```python
num_clusters = 6 #cluster의 갯수 산정
```

```python
kmeans_clustering = KMeans( n_clusters = num_clusters )#default
idx = kmeans_clustering.fit_predict( word_vectors ) #각 단어가 어떤 군집에 속하는지 확인
```

```python
# 각 어휘 단어를 클러스터 번호에 매핑되게 word/Index 사전을 만든다.
idx = list(idx)
names = model.wv.index2word
word_centroid_map = {names[i]: idx[i] for i in range(len(names))}#각 단어 별로 클러스터 번호에 매칭
```

```python
for cluster in range(0,6):
    # 클러스터 번호를 출력
    print("\nCluster {}".format(cluster))
    
    # 클러스터번호와 클러스터에 있는 단어를 찍는다.
    words = []
    for i in range(0,len(list(word_centroid_map.values()))): #list(dict)키 값만 리스트로 반환
        #list(dict.values)는 value값이 리스트로 반환
        if( list(word_centroid_map.values())[i] == cluster ):
            words.append(list(word_centroid_map.keys())[i])
    print(words)
```

```python
movie_info = pd.DataFrame({'movie_name':[],
                           'theater_spot':[],
                           'theater_info': [],
                           'theater_info2':[],
                           'sear':[],
                           'start_time':[],
                           'remain_seat':[],
                           'stop_time':[]})
#이제 데이터 프레임에 남은 좌석 보기 컬럼을 추가하여 클릭을하면 남은 좌석이 보이도록 
#img to url을 진행해야 하는데 흠...
for area in range(1,31):
    #len을 통해서 cgv가 있는 지역의 갯수를 산정 - 지역클릭
    browser.find_elements_by_xpath('//*[@class = "theater-select"]/div[2]/div/ul/li[1]/div/ul/li['+ str(area) + ']')[0].click()
    time.sleep(1.5)
    #len을 통해서 해당 cgv의 상영하는 날짜를 셀수도 있지만 지금은 토요일이라는 것으로 한정한다.
    #날짜 클릭
    #상영을 하지 않는 날짜는 조건문이나 예외처리를 통해서 상영을 하는 날짜만 클릭하게 한 뒤
    #그 뒤는 똑같이 크롤링
    #즉 영화장소와 날짜 사이에 조건문이나 예외처리 문이 들어가야함
    browser.find_elements_by_xpath('//*[@class = "date-list nano has-scrollbar has-scrollbar-y"]/ul/div/li[10]')[0].click()
    time.sleep(1.5)

    aven = browser.page_source
    aven2 = BeautifulSoup(aven,'html.parser')
    
    #영화이름
    movie_name = aven2.find('div',{'class':'movie-list nano has-scrollbar has-scrollbar-y'}).find('ul',{'class':'content scroll-y'}).find('li',{'class':'selected','selectedmovietype':'ALL'}).find('span',{'class':'text'}).text
    #극장위치
    theater_spot = aven2.find('div',{'class':'area_theater_list nano has-scrollbar has-scrollbar-y'}).find('li',{'class':'selected'}).find('a').text
    
    
    theater = aven2.find('div',{'class':'section section-time'}).find_all('div',{'class':'theater'})

    for i in range(len(theater)):
        theater_info = theater[i].find('span',{'class':'title'}).find('span',{'class':'name'}).text
        theater_info2 = theater[i].find('span',{'class':'title'}).find('span',{'class':'floor'}).text
        seat = theater[i].find('span',{'class':'title'}).find('span',{'class':'seatcount'}).text
        theater2 = theater[i].find_all('li')

        for j in range(len(theater2)):
            start_time = theater[i].find_all('li')[j].find('span',{'class':'time'}).text 
            remain_seat = theater[i].find_all('li')[j].find('span',{'class':'count'}).text
            stop_time = theater[i].find_all('li')[j].find('div',{'class':'sreader'}).text
            
            #영화별 어느 좌석이 남아있는지는 printsrc를 통해서 보여줌
            insert_data = pd.DataFrame({'movie_name':[movie_name],
                                       'theater_spot':[theater_spot],
                                       'theater_info': [theater_info],
                                       'theater_info2':[theater_info2],
                                       'sear':[seat],
                                       'start_time':[start_time],
                                       'remain_seat':[remain_seat],
                                       'stop_time':[stop_time]})
            movie_info = movie_info.append(insert_data)
```

```python
movie_info = pd.DataFrame({'movie_name':[],
                           'theater_spot':[],
                           'theater_info': [],
                           'theater_info2':[],
                           'sear':[],
                           'start_time':[],
                           'remain_seat':[],
                           'stop_time':[],
                           'ymd' : [],
                           'week': []})

for area in range(1,31):
    #len을 통해서 cgv가 있는 지역의 갯수를 산정 - 지역클릭
    #지역도 날짜처럼 상영하는 곳 안하는 곳 구분해야함
    browser.find_elements_by_xpath('//*[@class = "theater-select"]/div[2]/div/ul/li[1]/div/ul/li['+ str(area) + ']')[0].click()
    time.sleep(1.5)
    aven = browser.page_source
    aven2 = BeautifulSoup(aven,'html.parser')
    date_list0 = aven2.find('div',{'class':'date-list nano has-scrollbar has-scrollbar-y'}).find('ul',{'class':'content scroll-y'})
    date_list1 = date_list0.find_all('li')
    date_num = []
    for i in range(len(date_list1)):
        if 'month' in date_list1[i]['class']:
            pass
        elif 'day' in date_list1[i]['class']:
            if '선택불가' in date_list1[i].find('span',{'class':'sreader'}).text:
                pass
            else:
                date_num.append(i+1)        
        else:
            pass
    #len을 통해서 해당 cgv의 상영하는 날짜를 셀수도 있지만 지금은 토요일이라는 것으로 한정한다.
    #날짜 클릭
    time.sleep(2)
    for date2 in range(len(date_num)):
        try:
            browser.find_elements_by_xpath('//*[@class = "date-list nano has-scrollbar has-scrollbar-y"]/ul/div/li[' + str(date_num[date2]) + ']')[0].click()
            time.sleep(2)
            aven3 = browser.page_source
            aven4 = BeautifulSoup(aven3,'html.parser')
            ymd = date_list1[date_num[date2]-1]['date']
            week = date_list1[date_num[date2]-1].find('span',{'class':'dayweek'}).text
            #영화이름
            movie_name = aven4.find('div',{'class':'movie-list nano has-scrollbar has-scrollbar-y'}).find('ul',{'class':'content scroll-y'}).find('li',{'class':'selected','selectedmovietype':'ALL'}).find('span',{'class':'text'}).text
            #극장위치
            theater_spot = aven4.find('div',{'class':'area_theater_list nano has-scrollbar has-scrollbar-y'}).find('li',{'class':'selected'}).find('a').text


            theater = aven4.find('div',{'class':'section section-time'}).find_all('div',{'class':'theater'})

            for i in range(len(theater)):
                theater_info = theater[i].find('span',{'class':'title'}).find('span',{'class':'name'}).text
                theater_info2 = theater[i].find('span',{'class':'title'}).find('span',{'class':'floor'}).text
                seat = theater[i].find('span',{'class':'title'}).find('span',{'class':'seatcount'}).text
                theater2 = theater[i].find_all('li')

                for j in range(len(theater2)):
                    start_time = theater[i].find_all('li')[j].find('span',{'class':'time'}).text 
                    remain_seat = theater[i].find_all('li')[j].find('span',{'class':'count'}).text
                    stop_time = theater[i].find_all('li')[j].find('div',{'class':'sreader'}).text

                    #영화별 어느 좌석이 남아있는지는 printsrc를 통해서 보여줌
                    insert_data = pd.DataFrame({'movie_name':[movie_name],
                                               'theater_spot':[theater_spot],
                                               'theater_info': [theater_info],
                                               'theater_info2':[theater_info2],
                                               'sear':[seat],
                                               'start_time':[start_time],
                                               'remain_seat':[remain_seat],
                                               'stop_time':[stop_time],
                                               'ymd' : [ymd],
                                               'week': [week]})
                    movie_info = movie_info.append(insert_data)
        except:
            pass

movie_info.index = range(len(movie_info))




```

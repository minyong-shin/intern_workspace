## 실제 오차율비교 수정하면서 최적의 cm차이 찾기 v_0.3

모든 부위마다 멘트를 줄 때 음수가 되는 부분은 일단 무조건 1단계 멘트를 프론트에 주어주고  
백단에는 음수인부분 중에서도 - 어느정도까지는 허용할 수 있다는 설문조사한 결과를 토대로 그것을 기준으로 "입지 못한다" / "어느정도는 입을수 있다" 로 나눠서 분류해 놓는다.  

따라서 부위별로 상한만 구해서 상한의 percentage와 0%간의 구간을 5구간으로 나눠서 멘트를 준다.  

이때 최적의 cm를 찾기 위해서 오차율마다 내가 멘트가 마음에 드는지 여부를 확인하면서 진행한다.

계산식의문제는 없음
단순히 오차의 범위를 한정할 때 범위값을 좀 더 줄이면 좋을 것 같음 범위값이 크게 되면 거의 모든 상품이  
타이트, 슬림에서 멘트를 부여 받기 때문에 오차의 범위 값을 줄여야함.

오차의 범위를 줄이는 기준은 최대 하한으로 정하면 안될 것 같고 주관적으로 정해야 할 필요성이 있음.  
생각보다 오차의 범위를 엄청 줄여야 될 것 같은데..

```python
#a = 가슴 / b = 어깨 / c = 암홀 / d = 팔길이 / e = 총장  ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo(a,b,c,d,e,w1,w2,w3,w4,w5,data):
    #전역변수 처리하기
    num = data[data['name']=='스파오체크셔츠'].index.item() #num을 함수 밖에서 설정할 수 있음.
    
    #if '니트' in category_name:
    #if '셔츠' in category_name: 이처럼 함수 내부에서 카테고리별로 함수를 짜야함
    #옷의 카테고리마다 최대 하한을 정해야한다
    #-> 이유는 옷과 신체의 사이즈 차이가 (-) 어느정도가 된다면 아예 해당 사람한테 못입어버리는 옷이 되기 때문에 
    #최대 하한을 카테고리마다 정해준다면 user에게 그에 따른 조건문으로 추천을 해줄지 말지 정할 수 있음
    #또한 카테고리별로 하한을 따로 정해서 저장해야 하는 이유는 옷마다 신체와 옷의 차이가 어느정도 되었을때
    #못 입는 옷인지 다르기 때문에 카테고리별로 정해줘야함
    try:
        if abs(data['가슴'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['가슴'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임
        
        test = 100 - (-2/a)*100 #-2로했으면 무조건 단면으로 계산해야함
        
        if aa > 100: #가슴부위는 설문조사결과 -2cm까지는 커버가능
            if 100<aa<=test:
                print("back-end에 들어갈내용 : 이정도 차이까지는 입을 수 있음.")
            else:
                print("back-end에 들어갈내용 : 이 옷은 user에게 입지 못하는 옷.")       
        else:
            print("back-end에 들어갈 내용 : 옷이 신체보다 크므로 정상적으로 비교한다 :")
        #가슴은 최대 허용 오차가 55% 즉 fit percentage는 45% ~ 100%까지 허용가능 100% = 완전이 딱맞는것
        #percentage가 100이 넘는 것들은 무조건 1단계 값을 줘야함
        if aa==0:
            print("옷과 신체가 비정상적으로 차이남")
        elif 0<aa<45:
            print('error : 확인바람 ')
        elif 45<=aa<56:
            print('고객님에게 엄청 커요')
        elif 56<=aa<67:
            print('고객님에게 살짝 커요')
        elif 67<=aa<78:
            print('고객님에게 적당해요')
        elif 78<=aa<89:
            print('고객님에게 살짝 작아요')
        else:
            print('고객님에게 엄청 작아요')
        print(aa) 
        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('가슴부위의 사이즈가 없습니다.')
    
    try:
        if abs(data['어깨'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['어깨'].iloc[num]-b)/b)*100
        
        test2 = 100 - (-2/b)*100 #-2로했으면 무조건 단면으로 계산해야함

        if bb > 100: 
            if 100<bb<=test2:
                print("back-end에 들어갈내용 : 이정도 차이까지는 입을 수 있음.")
            else:
                print("back-end에 들어갈내용 : 이 옷은 user에게 입지 못하는 옷.")       
        else:
            print("back-end에 들어갈 내용 : 옷이 신체보다 크므로 정상적으로 비교한다 :")
        #어깨는 최대 허용 오차가 55% 즉 fit percentage는 48% ~ 100%까지 허용가능 100% = 완전이 딱맞는것
        #percentage가 100이 넘는 것들은 무조건 1단계 값을 줘야함 
        #암홀은 다른 부위와 다르게 굉장히 오버핏에도 많이 입는 경향이 있고 판매하는 경향이 있음 그래서
        #허용오차가 다른 부위에 비해서 굉장히 넓은 편이다.
        if bb==0:
            print('옷과 신체가 비정상적으로 차이남')
        elif 0<bb<48:
            print('error : 확인바람')
        elif 48<=bb<58.4:
            print('여유롭게 입을 수 있어요')
        elif 58.4<=bb<68.8:
            print('살짝 넉넉하게 입을 수 있어요')
        elif 68.8<=bb<79.2:
            print('고객님에게 적당해요')
        elif 79.2<=bb<89.6:
            print('슬림하게 입을 수 있어요')
        else:
            print('타이트하게 입을 수 있어요')
        print(bb)
        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('어깨부위의 사이즈가 없습니다.')
    
    try:
        if abs(data['암홀'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['암홀'].iloc[num]-c)/c)*100
        
        test3 = 100 - (-2/c)*100 #-2로했으면 무조건 단면으로 계산해야함

        if cc > 100: 
            if 100<cc<=test3:
                print("back-end에 들어갈내용 : 이정도 차이까지는 입을 수 있음.")
            else:
                print("back-end에 들어갈내용 : 이 옷은 user에게 입지 못하는 옷.")       
        else:
            print("back-end에 들어갈 내용 : 옷이 신체보다 크므로 정상적으로 비교한다 :")
        #어깨는 최대 허용 오차가 55% 즉 fit percentage는 2% ~ 100%까지 허용가능 100% = 완전이 딱맞는것
        #percentage가 100이 넘는 것들은 무조건 1단계 값을 줘야함 
        if cc==0:
            print('옷과 신체가 비정상적으로 차이남')
        elif 0<cc<2:
            print('error : 확인바람')
        elif 2<=cc<21.6:
            print('여유롭게 입을 수 있어요')
        elif 21.6<=cc<41.2:
            print('살짝 넉넉하게 입을 수 있어요')
        elif 41.2<=cc<60.8:
            print('고객님에게 적당해요')
        elif 60.8<=cc<80.4:
            print('슬림하게 입을 수 있어요')
        else:
            print('타이트하게 입을 수 있어요')  
        print(cc)  
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('암홀의 사이즈가 없습니다.')
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        if abs(data['팔길이'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['팔길이'].iloc[num]-d)/d)*100
        print('옷의 소매길이는 당신의 신체와 '+ str(round(data['팔길이'].iloc[num]-d,1)) + 'cm'+' 차이납니다.' )

        if dd > 100:
            dd = 200 - dd
        else:
            dd = dd
    except:
        print('팔길이의 사이즈가 없습니다.')
    
    try:
        if abs(data['총장'].iloc[num]-e) > e:
            ee=0 #오차가 심하다고하면 값을 0을 집어 넣어서 default를 맞춰놓자
        else:
            ee = 100 - ((data['총장'].iloc[num]-e)/e)*100
        print('옷의 총장은 당신의 신체와 '+ str(round(data['총장'].iloc[num]-e,1)) +'cm'+ ' 차이납니다.' )

        if ee > 100:
            ee = 200 - ee
        else:
            ee = ee
    except:
        print('총장의 사이즈가 없습니다.')
    #너무 오차가 많이 나게 된다면 음수값이 나올수가 있기 때문에
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4 + ee*w5)/(w1+w2+w3+w4+w5)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 80 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 90 % 부합합니다'
    else:
        return '당신의 신체사이즈와 98 % 부합합니다'
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님
```

# 오차의 범위를 대폭 줄인 알고리즘 v_0.4

오차의 범위를 대폭 줄이지 않고 각 부위별로 상한과 하한을 정해서 그 범위를 5구간으로 나눠서 부위별  
멘트를 주게 되면 거의 모든 옷들이 엄청작아요라는 1단계 멘트를 부여 받게 된다 따라서 상한과 하한으로  
구간을 나눠 멘트를 부여하지 않고 임의로 하한을 부여하여 (v_0.4는 75%~100%로 정함) 5구간으로 나누면  
어느정도 내 신체와 비슷하게 멘트를 부여 받을 수 있다.

다른 부위는 올바른 멘트를 부여 받지만 어깨부분만 1단계 정도 이상하게 부여 받기 때문에   
어깨부분의 오차만 수정하면 될 것 같음

```python
data = pd.DataFrame({'name':['자두십자수니트','바나나상의','스파오체크셔츠','발렌티노리치'],
                    '가슴':[63,68,61,60.5],
                    '어깨':[45,47,51.5,53],
                    '암홀':[25,31,25,25],
                    '팔길이':[50,45,'null',60],
                    '총장':[72,69,80,82]})
#직접 null값을 줘서 결측치 부여 가능
```

```python
data = data.replace(data.iloc[2][4:],'null') #데이터 프레임에 널값부여하는 것
```

```python
#a = 가슴 / b = 어깨 / c = 암홀 / d = 팔길이 / e = 총장  ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo(a,b,c,d,e,w1,w2,w3,w4,w5,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    top_maximum_err = -2 #상의 최대 허용 음수 오차
    error_intrv0 =75 #오차구간
    error_intrv1 =80
    error_intrv2 =85
    error_intrv3 =90
    error_intrv4 =95 
    #if '니트' in category_name:
    #if '셔츠' in category_name: 이처럼 함수 내부에서 카테고리별로 함수를 짜야함
    #옷의 카테고리마다 최대 하한을 정해야한다
    #-> 이유는 옷과 신체의 사이즈 차이가 (-) 어느정도가 된다면 아예 해당 사람한테 못입어버리는 옷이 되기 때문에 
    #최대 하한을 카테고리마다 정해준다면 user에게 그에 따른 조건문으로 추천을 해줄지 말지 정할 수 있음
    #또한 카테고리별로 하한을 따로 정해서 저장해야 하는 이유는 옷마다 신체와 옷의 차이가 어느정도 되었을때
    #못 입는 옷인지 다르기 때문에 카테고리별로 정해줘야함
    try:
        if abs(data['가슴'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['가슴'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임
        
        test = 100 - (top_maximum_err/a)*100 #-2로했으면 무조건 단면으로 계산해야함
        
        if aa > 100: #가슴부위는 설문조사결과 -2cm까지는 커버가능
            if 100<aa<=test:
                print("back-end에 들어갈내용 : 이정도 차이까지는 입을 수 있음.")
            else:
                print("back-end에 들어갈내용 : 이 옷은 user에게 입지 못하는 옷.")       
        else:
            print("back-end에 들어갈 내용 : 옷이 신체보다 크므로 정상적으로 비교한다 :")
        #가슴은 최대 허용 오차가 55% 즉 fit percentage는 45% ~ 100%까지 허용가능 100% = 완전이 딱맞는것
        #percentage가 100이 넘는 것들은 무조건 1단계 값을 줘야함
        if aa==0:
            print("옷과 신체가 비정상적으로 차이남")
        elif 0<aa<error_intrv0:
            print('error : 확인바람 ')
        elif error_intrv0<=aa<error_intrv1:
            print('고객님에게 엄청 커요')
        elif error_intrv1<=aa<error_intrv2:
            print('고객님에게 살짝 커요')
        elif error_intrv2<=aa<error_intrv3:
            print('고객님에게 적당해요')
        elif error_intrv3<=aa<error_intrv4:
            print('고객님에게 살짝 작아요')
        else:
            print('고객님에게 엄청 작아요')#옷이 신체보다 작은 범위 즉, 음수의 범위까지 포괄
        print(aa) 
        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('가슴부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        if abs(data['어깨'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['어깨'].iloc[num]-b)/b)*100
        
        test2 = 100 - (top_maximum_err/b)*100 #-2로했으면 무조건 단면으로 계산해야함

        if bb > 100: 
            if 100<bb<=test2:
                print("back-end에 들어갈내용 : 이정도 차이까지는 입을 수 있음.")
            else:
                print("back-end에 들어갈내용 : 이 옷은 user에게 입지 못하는 옷.")       
        else:
            print("back-end에 들어갈 내용 : 옷이 신체보다 크므로 정상적으로 비교한다 :")
        #어깨는 최대 허용 오차가 55% 즉 fit percentage는 48% ~ 100%까지 허용가능 100% = 완전이 딱맞는것
        #percentage가 100이 넘는 것들은 무조건 1단계 값을 줘야함 
        #암홀은 다른 부위와 다르게 굉장히 오버핏에도 많이 입는 경향이 있고 판매하는 경향이 있음 그래서
        #허용오차가 다른 부위에 비해서 굉장히 넓은 편이다.
        if bb==0:
            print('옷과 신체가 비정상적으로 차이남')
        elif 0<bb<error_intrv0:
            print('error : 확인바람')
        elif error_intrv0<=bb<error_intrv1:
            print('여유롭게 입을 수 있어요')
        elif error_intrv1<=bb<error_intrv2:
            print('살짝 넉넉하게 입을 수 있어요')
        elif error_intrv2<=bb<error_intrv3:
            print('고객님에게 적당해요')
        elif error_intrv3<=bb<error_intrv4:
            print('슬림하게 입을 수 있어요')
        else:
            print('타이트하게 입을 수 있어요')
        print(bb)
        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('어깨부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        if abs(data['암홀'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['암홀'].iloc[num]-c)/c)*100
        
        test3 = 100 - (top_maximum_err/c)*100 #-2로했으면 무조건 단면으로 계산해야함

        if cc > 100: 
            if 100<cc<=test3:
                print("back-end에 들어갈내용 : 이정도 차이까지는 입을 수 있음.")
            else:
                print("back-end에 들어갈내용 : 이 옷은 user에게 입지 못하는 옷.")       
        else:
            print("back-end에 들어갈 내용 : 옷이 신체보다 크므로 정상적으로 비교한다 :")
        #어깨는 최대 허용 오차가 55% 즉 fit percentage는 2% ~ 100%까지 허용가능 100% = 완전이 딱맞는것
        #percentage가 100이 넘는 것들은 무조건 1단계 값을 줘야함 
        if cc==0:
            print('옷과 신체가 비정상적으로 차이남')
        elif 0<cc<error_intrv0:
            print('error : 확인바람')
        elif error_intrv0<=cc<error_intrv1:
            print('여유롭게 입을 수 있어요')
        elif error_intrv1<=cc<error_intrv2:
            print('살짝 넉넉하게 입을 수 있어요')
        elif error_intrv2<=cc<error_intrv3:
            print('고객님에게 적당해요')
        elif error_intrv3<=cc<error_intrv4:
            print('슬림하게 입을 수 있어요')
        else:
            print('타이트하게 입을 수 있어요')  
        print(cc)  
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('암홀의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        if abs(data['팔길이'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['팔길이'].iloc[num]-d)/d)*100
        print('옷의 소매길이는 당신의 신체와 '+ str(round(data['팔길이'].iloc[num]-d,1)) + 'cm'+' 차이납니다.' )

        if dd > 100:
            dd = 200 - dd
        else:
            dd = dd
    except:
        print('팔길이의 사이즈가 없습니다.')
        dd = 0
        w4 = 0
    
    try:
        if abs(data['총장'].iloc[num]-e) > e:
            ee=0 #오차가 심하다고하면 값을 0을 집어 넣어서 default를 맞춰놓자
        else:
            ee = 100 - ((data['총장'].iloc[num]-e)/e)*100
        print('옷의 총장은 당신의 신체와 '+ str(round(data['총장'].iloc[num]-e,1)) +'cm'+ ' 차이납니다.' )

        if ee > 100:
            ee = 200 - ee
        else:
            ee = ee
    except:
        print('총장의 사이즈가 없습니다.')
        ee=0
        w5=0
    #너무 오차가 많이 나게 된다면 음수값이 나올수가 있기 때문에
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4 + ee*w5)/(w1+w2+w3+w4+w5)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 80 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 90 % 부합합니다'
    else:
        return '당신의 신체사이즈와 98 % 부합합니다'
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님
```

```python
#a = 가슴 / b = 어깨 / c = 암홀 / d = 팔길이 / e = 총장  ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo(a,b,c,d,e,w1,w2,w3,w4,w5,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    #아예 우리가 잘못알고 있었음 -> 신체보다 작아도 입을 수 있음
    top_maximum_err = -2 #상의 최대 허용 음수 오차
    #어깨
    shoulder_intrv0 =b-4 #오차구간
    shoulder_intrv1 = b-2
    shoulder_intrv2 = b+3 # 반올림을 해야하는지
    shoulder_intrv3 = b+6
    
    #암홀
    armhole_intrv0 = c-3
    armhole_intrv1 = c-1.5
    armhole_intrv2 = c+2
    armhole_intrv3 = c+4
    
    #가슴
    chest_intrv0 = a-3
    chest_intrv1 = a-1.5
    chest_intrv2 = a+3.5
    chest_intrv3 = a+7
    
    
    #if '니트' in category_name:
    #if '셔츠' in category_name: 이처럼 함수 내부에서 카테고리별로 함수를 짜야함
    #옷의 카테고리마다 최대 하한을 정해야한다
    #-> 이유는 옷과 신체의 사이즈 차이가 (-) 어느정도가 된다면 아예 해당 사람한테 못입어버리는 옷이 되기 때문에 
    #최대 하한을 카테고리마다 정해준다면 user에게 그에 따른 조건문으로 추천을 해줄지 말지 정할 수 있음
    #또한 카테고리별로 하한을 따로 정해서 저장해야 하는 이유는 옷마다 신체와 옷의 차이가 어느정도 되었을때
    #못 입는 옷인지 다르기 때문에 카테고리별로 정해줘야함
    try:
        print(str(data['가슴'].iloc[num] - a)+'cm' '차이납니다')
        chest_num = data['가슴'].iloc[num] 
        
        if chest_num <=chest_intrv0:
            print("매우 타이트")
        elif chest_intrv0 < chest_num <= chest_intrv1:
            print("타이트")
        elif chest_intrv1 < chest_num <= chest_intrv2:
            print("적당")
        elif chest_intrv2 < chest_num <= chest_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['가슴'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['가슴'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임
        
        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('가슴부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        print(str(data['어깨'].iloc[num] - b)+'cm' '차이납니다')
        shoulder_num = data['어깨'].iloc[num] 
        
        if shoulder_num <=shoulder_intrv0:
            print("매우 타이트")
        elif shoulder_intrv0 < shoulder_num <= shoulder_intrv1:
            print("타이트")
        elif shoulder_intrv1 < shoulder_num <= shoulder_intrv2:
            print("적당")
        elif shoulder_intrv2 < shoulder_num <= shoulder_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['어깨'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['어깨'].iloc[num]-b)/b)*100
        
        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('어깨부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        print(str(data['암홀'].iloc[num] - c)+'cm' '차이납니다')
        armhole_num = data['암홀'].iloc[num] 
        
        if armhole_num <=armhole_intrv0:
            print("매우 타이트")
        elif armhole_intrv0 < armhole_num <= armhole_intrv1:
            print("타이트")
        elif armhole_intrv1 < armhole_num <= armhole_intrv2:
            print("적당")
        elif armhole_intrv2 < armhole_num <= armhole_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['암홀'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['암홀'].iloc[num]-c)/c)*100
        
        
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('암홀의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        if abs(data['팔길이'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['팔길이'].iloc[num]-d)/d)*100
        print('옷의 소매길이는 당신의 신체와 '+ str(round(data['팔길이'].iloc[num]-d,1)) + 'cm'+' 차이납니다.' )

        if dd > 100:
            dd = 200 - dd
        else:
            dd = dd
    except:
        print('팔길이의 사이즈가 없습니다.')
        dd = 0
        w4 = 0
    
    try:
        if abs(data['총장'].iloc[num]-e) > e:
            ee=0 #오차가 심하다고하면 값을 0을 집어 넣어서 default를 맞춰놓자
        else:
            ee = 100 - ((data['총장'].iloc[num]-e)/e)*100
        print('옷의 총장은 당신의 신체와 '+ str(round(data['총장'].iloc[num]-e,1)) +'cm'+ ' 차이납니다.' )

        if ee > 100:
            ee = 200 - ee
        else:
            ee = ee
    except:
        print('총장의 사이즈가 없습니다.')
        ee=0
        w5=0
        
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4 + ee*w5)/(w1+w2+w3+w4+w5)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 80 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 90 % 부합합니다'
    else:
        return '당신의 신체사이즈와 98 % 부합합니다'
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님
```

```python
#a = 가슴 / b = 어깨 / c = 암홀 / d = 팔길이 / e = 총장  ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo_m_pants(a,b,c,d,w1,w2,w3,w4,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    #아예 우리가 잘못알고 있었음 -> 신체보다 작아도 입을 수 있음
    #top_maximum_err = -2 #상의 최대 허용 음수 오차
    #허리
    waist_intrv0 =b-2 #오차구간
    waist_intrv1 = b-1
    waist_intrv2 = b+2
    waist_intrv3 = b+4
    
    #엉덩이
    hip_intrv0 = c-2
    hip_intrv1 = c-1
    hip_intrv2 = c+2
    hip_intrv3 = c+4
    
    #허벅지
    thigh_intrv0 = a-2
    thigh_intrv1 = a-1
    thigh_intrv2 = a+1.5
    thigh_intrv3 = a+3
    
    
    #if '니트' in category_name:
    #if '셔츠' in category_name: 이처럼 함수 내부에서 카테고리별로 함수를 짜야함
    #옷의 카테고리마다 최대 하한을 정해야한다
    #-> 이유는 옷과 신체의 사이즈 차이가 (-) 어느정도가 된다면 아예 해당 사람한테 못입어버리는 옷이 되기 때문에 
    #최대 하한을 카테고리마다 정해준다면 user에게 그에 따른 조건문으로 추천을 해줄지 말지 정할 수 있음
    #또한 카테고리별로 하한을 따로 정해서 저장해야 하는 이유는 옷마다 신체와 옷의 차이가 어느정도 되었을때
    #못 입는 옷인지 다르기 때문에 카테고리별로 정해줘야함
    
    #멘트부여
    try:
        print(str(data['허벅지'].iloc[num] - a)+'cm' '차이납니다')
        thigh_num = data['허벅지'].iloc[num] 
        
        if thigh_num <=thigh_intrv0:
            print("매우 타이트")
        elif thigh_intrv0 < thigh_num <= thigh_intrv1:
            print("타이트")
        elif thigh_intrv1 < thigh_num <= thigh_intrv2:
            print("적당")
        elif thigh_intrv2 < thigh_num <= thigh_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['허벅지'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['허벅지'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임
        
        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('허벅지부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        print(str(data['허리'].iloc[num] - b)+'cm' '차이납니다')
        waist_num = data['허리'].iloc[num] 
        
        if waist_num <=waist_intrv0:
            print("매우 타이트")
        elif waist_intrv0 < waist_num <= waist_intrv1:
            print("타이트")
        elif waist_intrv1 < waist_num <= waist_intrv2:
            print("적당")
        elif waist_intrv2 < waist_num <= waist_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['허리'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['허리'].iloc[num]-b)/b)*100
        
        
        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('허리부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        print(str(data['엉덩이'].iloc[num] - c)+'cm' '차이납니다')
        hip_num = data['엉덩이'].iloc[num] 
        
        if hip_num <=hip_intrv0:
            print("매우 타이트")
        elif hip_intrv0 < hip_num <= hip_intrv1:
            print("타이트")
        elif hip_intrv1 < hip_num <= hip_intrv2:
            print("적당")
        elif hip_intrv2 < hip_num <= hip_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['엉덩이'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['엉덩이'].iloc[num]-c)/c)*100
        
        
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('엉덩이의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        if abs(data['바지총장'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['바지총장'].iloc[num]-d)/d)*100
        print('옷의 바지 총장은 당신의 신체와 '+ str(round(data['바지총장'].iloc[num]-d,1)) + 'cm'+' 차이납니다.' )

        if dd > 100:
            dd = 200 - dd
        else:
            dd = dd
    except:
        print('바지총장의 사이즈가 없습니다.')
        dd = 0
        w4 = 0
    
    
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4)/(w1+w2+w3+w4)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 65 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 75 % 부합합니다'
    else:
        return '당신의 신체사이즈와' + str(round(mean,1)) + '% 부합합니다.' 
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님

#어차피 백단에 들어가는 내용이겠지만
#신체사이즈 퍼센티지를 80%까지는 percentage그대로 보여주고 80%보다 아래의percentage를 받았다면
#그때는 조건문을 넣어서 60%~80%를 작위적으로 받도록하자
#이유는 오버핏을 제외하고는 신체와 그렇게 차이나는 경우가 없어 사이즈가 있는 모든 옷들은 왠만해서는
#80%이상의 신체 percentage를 가진다 하지만 오버핏인 경우에만 80%아래로 내려가는 경우 혹은
#말도안되게 옷과 신체치수가 차이나는 경우에 80%아래로 나는 경우가 존재하므로 이와 같은 방법을
#이용하면 좋을 듯
```

```python
#a = 어깨 / b = 가슴 / c = 암홀 / d = 허리 / e = 엉덩이 /f = 총장 ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo_m_outer(a,b,c,d,e,f,w1,w2,w3,w4,w5,w6,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    #아예 우리가 잘못알고 있었음 -> 신체보다 작아도 입을 수 있음
    #top_maximum_err = -2 #상의 최대 허용 음수 오차

    #어깨
    shoulder_intrv0 = a+1
    shoulder_intrv1 = a+2
    shoulder_intrv2 = a+7.5
    shoulder_intrv3 = a+11
    
    #가슴
    chest_intrv0 =b+2 #오차구간
    chest_intrv1 = b+4.5
    chest_intrv2 = b+9.5
    chest_intrv3 = b+13.5
    
    #암홀
    armhole_intrv0 = c+1.5
    armhole_intrv1 = c+3
    armhole_intrv2 = c+7
    armhole_intrv3 = c+10.5
    
    #허리
    waist_intrv0 =d+6 #오차구간
    waist_intrv1 = d+12
    waist_intrv2 = d+18.5
    waist_intrv3 = d+27.5
    
    #엉덩이
    hip_intrv0 = e+3.5
    hip_intrv1 = e+7
    hip_intrv2 = e+11.5
    hip_intrv3 = e+17
    

    
    #멘트부여
    try:
        print(str(data['어깨'].iloc[num] - a)+'cm' '차이납니다')
        shoulder_num = data['어깨'].iloc[num] 
        
        if shoulder_num <=shoulder_intrv0:
            print("매우 타이트")
        elif shoulder_intrv0 < shoulder_num <= shoulder_intrv1:
            print("타이트")
        elif shoulder_intrv1 < shoulder_num <= shoulder_intrv2:
            print("적당")
        elif shoulder_intrv2 < shoulder_num <= shoulder_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['어깨'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['어깨'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임

        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('어깨부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        print(str(data['가슴'].iloc[num] - b)+'cm' '차이납니다')
        chest_num = data['가슴'].iloc[num] 
        
        if chest_num <=chest_intrv0:
            print("매우 타이트")
        elif chest_intrv0 < chest_num <= chest_intrv1:
            print("타이트")
        elif chest_intrv1 < chest_num <= chest_intrv2:
            print("적당")
        elif chest_intrv2 < chest_num <= chest_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['가슴'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['가슴'].iloc[num]-b)/b)*100

        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('가슴부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        print(str(data['암홀'].iloc[num] - c)+'cm' '차이납니다')
        armhole_num = data['암홀'].iloc[num] 
        
        if armhole_num <=armhole_intrv0:
            print("매우 타이트")
        elif armhole_intrv0 < armhole_num <= armhole_intrv1:
            print("타이트")
        elif armhole_intrv1 < armhole_num <= armhole_intrv2:
            print("적당")
        elif armhole_intrv2 < armhole_num <= armhole_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['암홀'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['암홀'].iloc[num]-c)/c)*100
        
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('암홀의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        print(str(data['허리'].iloc[num] - d)+'cm' '차이납니다')
        waist_num = data['허리'].iloc[num] 
        
        if waist_num <=waist_intrv0:
            print("매우 타이트")
        elif waist_intrv0 < waist_num <= waist_intrv1:
            print("타이트")
        elif waist_intrv1 < waist_num <= waist_intrv2:
            print("적당")
        elif waist_intrv2 < waist_num <= waist_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['허리'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['허리'].iloc[num]-d)/d)*100
        
        if dd>100:
            dd= 200 - dd
        else:
            dd = dd
    except:
        print('암홀의 사이즈가 없습니다.')
        dd=0
        w4=0
        
    try:
        print(str(data['엉덩이'].iloc[num] - e)+'cm' '차이납니다')
        hip_num = data['엉덩이'].iloc[num] 
        
        if hip_num <=hip_intrv0:
            print("매우 타이트")
        elif hip_intrv0 < hip_num <= hip_intrv1:
            print("타이트")
        elif hip_intrv1 < hip_num <= hip_intrv2:
            print("적당")
        elif hip_intrv2 < hip_num <= hip_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['엉덩이'].iloc[num]-e) > e:
            ee=0
        else:
            ee = 100 - ((data['엉덩이'].iloc[num]-e)/e)*100
        
        if ee>100:
            ee= 200 - ee
        else:
            ee = ee
    except:
        print('암홀의 사이즈가 없습니다.')
        ee=0
        w5=0
        
    try:
        if abs(data['바지총장'].iloc[num]-f) > f:
            ff=0
        else:
            ff = 100 - ((data['바지총장'].iloc[num]-f)/f)*100
        print('아우터의 총장은 당신의 신체와 '+ str(round(data['바지총장'].iloc[num]-f,1)) + 'cm'+' 차이납니다.' )

        if ff > 100:
            ff = 200 - ff
        else:
            ff = ff
    except:
        print('총장의 사이즈가 없습니다.')
        ff = 0
        w6 = 0
    
    #너무 오차가 많이 나게 된다면 음수값이 나올수가 있기 때문에
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4 + ee*w5 + ff*w6)/(w1+w2+w3+w4+w5+w6)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 65 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 75 % 부합합니다'
    else:
        return '당신의 신체사이즈와' + str(round(mean,1)) + '% 부합합니다.' 
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님

#어차피 백단에 들어가는 내용이겠지만
#신체사이즈 퍼센티지를 80%까지는 percentage그대로 보여주고 80%보다 아래의percentage를 받았다면
#그때는 조건문을 넣어서 60%~80%를 작위적으로 받도록하자
#이유는 오버핏을 제외하고는 신체와 그렇게 차이나는 경우가 없어 사이즈가 있는 모든 옷들은 왠만해서는
#80%이상의 신체 percentage를 가진다 하지만 오버핏인 경우에만 80%아래로 내려가는 경우 혹은
#말도안되게 옷과 신체치수가 차이나는 경우에 80%아래로 나는 경우가 존재하므로 이와 같은 방법을
#이용하면 좋을 듯
```

```python
#a = 가슴 / b = 어깨 / c = 암홀 / d = 팔길이 / e = 총장  ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo_w(a,b,c,d,e,w1,w2,w3,w4,w5,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    #아예 우리가 잘못알고 있었음 -> 신체보다 작아도 입을 수 있음
    top_maximum_err = -2 #상의 최대 허용 음수 오차
    #어깨
    shoulder_intrv0 =b-4 #오차구간
    shoulder_intrv1 = b-2
    shoulder_intrv2 = b+2
    shoulder_intrv3 = b+4
    
    #암홀
    armhole_intrv0 = c-1
    armhole_intrv1 = c-0.5
    armhole_intrv2 = c+2
    armhole_intrv3 = c+4
    
    #가슴
    chest_intrv0 = a-2
    chest_intrv1 = a-1
    chest_intrv2 = a+5
    chest_intrv3 = a+10
    
    
    #if '니트' in category_name:
    #if '셔츠' in category_name: 이처럼 함수 내부에서 카테고리별로 함수를 짜야함
    #옷의 카테고리마다 최대 하한을 정해야한다
    #-> 이유는 옷과 신체의 사이즈 차이가 (-) 어느정도가 된다면 아예 해당 사람한테 못입어버리는 옷이 되기 때문에 
    #최대 하한을 카테고리마다 정해준다면 user에게 그에 따른 조건문으로 추천을 해줄지 말지 정할 수 있음
    #또한 카테고리별로 하한을 따로 정해서 저장해야 하는 이유는 옷마다 신체와 옷의 차이가 어느정도 되었을때
    #못 입는 옷인지 다르기 때문에 카테고리별로 정해줘야함
    try:
        print(str(data['가슴'].iloc[num] - a)+'cm' '차이납니다')
        chest_num = data['가슴'].iloc[num] 
        
        if chest_num <=chest_intrv0:
            print("매우 타이트")
        elif chest_intrv0 < chest_num <= chest_intrv1:
            print("타이트")
        elif chest_intrv1 < chest_num <= chest_intrv2:
            print("적당")
        elif chest_intrv2 < chest_num <= chest_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['가슴'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['가슴'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임
        
        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('가슴부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        print(str(data['어깨'].iloc[num] - b)+'cm' '차이납니다')
        shoulder_num = data['어깨'].iloc[num] 
        
        if shoulder_num <=shoulder_intrv0:
            print("매우 타이트")
        elif shoulder_intrv0 < shoulder_num <= shoulder_intrv1:
            print("타이트")
        elif shoulder_intrv1 < shoulder_num <= shoulder_intrv2:
            print("적당")
        elif shoulder_intrv2 < shoulder_num <= shoulder_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['어깨'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['어깨'].iloc[num]-b)/b)*100
        
        
        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('어깨부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        print(str(data['암홀'].iloc[num] - c)+'cm' '차이납니다')
        armhole_num = data['암홀'].iloc[num] 
        
        if armhole_num <=armhole_intrv0:
            print("매우 타이트")
        elif armhole_intrv0 < armhole_num <= armhole_intrv1:
            print("타이트")
        elif armhole_intrv1 < armhole_num <= armhole_intrv2:
            print("적당")
        elif armhole_intrv2 < armhole_num <= armhole_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            

        if abs(data['암홀'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['암홀'].iloc[num]-c)/c)*100
        
        
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('암홀의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        if abs(data['팔길이'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['팔길이'].iloc[num]-d)/d)*100
        print('옷의 소매길이는 당신의 신체와 '+ str(round(data['팔길이'].iloc[num]-d,1)) + 'cm'+' 차이납니다.' )

        if dd > 100:
            dd = 200 - dd
        else:
            dd = dd
    except:
        print('팔길이의 사이즈가 없습니다.')
        dd = 0
        w4 = 0
    
    try:
        if abs(data['총장'].iloc[num]-e) > e:
            ee=0 #오차가 심하다고하면 값을 0을 집어 넣어서 default를 맞춰놓자
        else:
            ee = 100 - ((data['총장'].iloc[num]-e)/e)*100
        print('옷의 총장은 당신의 신체와 '+ str(round(data['총장'].iloc[num]-e,1)) +'cm'+ ' 차이납니다.' )

        if ee > 100:
            ee = 200 - ee
        else:
            ee = ee
    except:
        print('총장의 사이즈가 없습니다.')
        ee=0
        w5=0
    
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4+ ee*w5)/(w1+w2+w3+w4+w5)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 65 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 75 % 부합합니다'
    else:
        return '당신의 신체사이즈와 ' + str(round(mean,1)) + '% 부합합니다.' 
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님

#어차피 백단에 들어가는 내용이겠지만
#신체사이즈 퍼센티지를 80%까지는 percentage그대로 보여주고 80%보다 아래의percentage를 받았다면
#그때는 조건문을 넣어서 60%~80%를 작위적으로 받도록하자
#이유는 오버핏을 제외하고는 신체와 그렇게 차이나는 경우가 없어 사이즈가 있는 모든 옷들은 왠만해서는
#80%이상의 신체 percentage를 가진다 하지만 오버핏인 경우에만 80%아래로 내려가는 경우 혹은
#말도안되게 옷과 신체치수가 차이나는 경우에 80%아래로 나는 경우가 존재하므로 이와 같은 방법을
#이용하면 좋을 듯
```

```python
#a = 허벅지 / b = 허리 / c = 엉덩이 / d = 바지총장  ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo_w_pants(a,b,c,d,w1,w2,w3,w4,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    #아예 우리가 잘못알고 있었음 -> 신체보다 작아도 입을 수 있음
    #top_maximum_err = -2 #상의 최대 허용 음수 오차
    #허리
    waist_intrv0 =b-2 #오차구간
    waist_intrv1 = b-1
    waist_intrv2 = b+1
    waist_intrv3 = b+2
    
    #엉덩이
    hip_intrv0 = c-1
    hip_intrv1 = c-0.5
    hip_intrv2 = c+1.5
    hip_intrv3 = c+3
    
    #허벅지
    thigh_intrv0 = a-2
    thigh_intrv1 = a-1
    thigh_intrv2 = a+1
    thigh_intrv3 = a+2
    
    
    #if '니트' in category_name:
    #if '셔츠' in category_name: 이처럼 함수 내부에서 카테고리별로 함수를 짜야함
    #옷의 카테고리마다 최대 하한을 정해야한다
    #-> 이유는 옷과 신체의 사이즈 차이가 (-) 어느정도가 된다면 아예 해당 사람한테 못입어버리는 옷이 되기 때문에 
    #최대 하한을 카테고리마다 정해준다면 user에게 그에 따른 조건문으로 추천을 해줄지 말지 정할 수 있음
    #또한 카테고리별로 하한을 따로 정해서 저장해야 하는 이유는 옷마다 신체와 옷의 차이가 어느정도 되었을때
    #못 입는 옷인지 다르기 때문에 카테고리별로 정해줘야함
    try:
        print(str(data['허벅지'].iloc[num] - a)+'cm 차이납니다')
        thigh_num = data['허벅지'].iloc[num] 
        
        if thigh_num <=thigh_intrv0:
            print("매우 타이트")
        elif thigh_intrv0 < thigh_num <= thigh_intrv1:
            print("타이트")
        elif thigh_intrv1 < thigh_num <= thigh_intrv2:
            print("적당")
        elif thigh_intrv2 < thigh_num <= thigh_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['허벅지'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['허벅지'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임
        
        
        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('허벅지부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        print(str(data['허리'].iloc[num] - b)+'cm 차이납니다')
        waist_num = data['허리'].iloc[num] 
        
        if waist_num <=waist_intrv0:
            print("매우 타이트")
        elif waist_intrv0 < waist_num <= waist_intrv1:
            print("타이트")
        elif waist_intrv1 < waist_num <= waist_intrv2:
            print("적당")
        elif waist_intrv2 < waist_num <= waist_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['허리'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['허리'].iloc[num]-b)/b)*100
        
        
        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('허리부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        print(str(data['엉덩이'].iloc[num] - c)+'cm 차이납니다')
        hip_num = data['엉덩이'].iloc[num] 
        
        if hip_num <=hip_intrv0:
            print("매우 타이트")
        elif hip_intrv0 < hip_num <= hip_intrv1:
            print("타이트")
        elif hip_intrv1 < hip_num <= hip_intrv2:
            print("적당")
        elif hip_intrv2 < hip_num <= hip_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['엉덩이'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['엉덩이'].iloc[num]-c)/c)*100
        
        
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('엉덩이의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        if abs(data['바지총장'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['바지총장'].iloc[num]-d)/d)*100
        print('옷의 바지 총장은 당신의 신체와 '+ str(round(data['바지총장'].iloc[num]-d,1)) + 'cm'+' 차이납니다.' )

        if dd > 100:
            dd = 200 - dd
        else:
            dd = dd
    except:
        print('바지총장의 사이즈가 없습니다.')
        dd = 0
        w4 = 0
    
    #너무 오차가 많이 나게 된다면 음수값이 나올수가 있기 때문에
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4)/(w1+w2+w3+w4)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 65 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 75 % 부합합니다'
    else:
        return '당신의 신체사이즈와' + str(round(mean,1)) + '% 부합합니다.' 
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님

#어차피 백단에 들어가는 내용이겠지만
#신체사이즈 퍼센티지를 80%까지는 percentage그대로 보여주고 80%보다 아래의percentage를 받았다면
#그때는 조건문을 넣어서 60%~80%를 작위적으로 받도록하자
#이유는 오버핏을 제외하고는 신체와 그렇게 차이나는 경우가 없어 사이즈가 있는 모든 옷들은 왠만해서는
#80%이상의 신체 percentage를 가진다 하지만 오버핏인 경우에만 80%아래로 내려가는 경우 혹은
#말도안되게 옷과 신체치수가 차이나는 경우에 80%아래로 나는 경우가 존재하므로 이와 같은 방법을
#이용하면 좋을 듯
```

```python
#a = 어깨 / b = 가슴 / c = 암홀 / d = 허리 / e = 엉덩이 /f = 총장 ==> 신체치수
#여기서 성별/키,몸무게별 default로 들어온 a,b,c,d,e를 조정할 수 있을 것 같음 아침에 말한 +3cm
#지금은 임의로 옷의 값을 집어 넣었지만 본격적으로 집어 넣을 때는 데이터프레임 형식을 집어넣어서 필요한
#컬럼과 행을 뽑아서 사용할 것임.
def algo_w_outer(a,b,c,d,e,f,w1,w2,w3,w4,w5,w6,data,item_name):
    #전역변수 처리하기
    num = data[data['name']==item_name].index.item() #num을 함수 밖에서 설정할 수 있음.
    #전역변수 설정
    #아예 우리가 잘못알고 있었음 -> 신체보다 작아도 입을 수 있음
    #top_maximum_err = -2 #상의 최대 허용 음수 오차

    #어깨
    shoulder_intrv0 = a-6
    shoulder_intrv1 = a-4
    shoulder_intrv2 = a+1
    shoulder_intrv3 = a+2
    
    #가슴
    chest_intrv0 =b+3 #오차구간
    chest_intrv1 = b+6
    chest_intrv2 = b+11
    chest_intrv3 = b+16
    
    #암홀
    armhole_intrv0 = c+2
    armhole_intrv1 = c+4
    armhole_intrv2 = c+7
    armhole_intrv3 = c+10.5
    
    #허리
    waist_intrv0 =d+3 #오차구간
    waist_intrv1 = d+5.5
    waist_intrv2 = d+7
    waist_intrv3 = d+10.5
    
    #엉덩이
    hip_intrv0 = e+0.5
    hip_intrv1 = e+1.5
    hip_intrv2 = e+6.5
    hip_intrv3 = e+9.5
    

    
    #멘트부여
    try:
        print(str(data['어깨'].iloc[num] - a)+'cm' '차이납니다')
        shoulder_num = data['어깨'].iloc[num] 
        
        if shoulder_num <=shoulder_intrv0:
            print("매우 타이트")
        elif shoulder_intrv0 < shoulder_num <= shoulder_intrv1:
            print("타이트")
        elif shoulder_intrv1 < shoulder_num <= shoulder_intrv2:
            print("적당")
        elif shoulder_intrv2 < shoulder_num <= shoulder_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['어깨'].iloc[num]-a) > a:
            aa=0
        else: #옷의 사이즈 - 신체 사이즈
            aa = 100 - ((data['어깨'].iloc[num]-a)/a)*100 #해당 구하는 것은 1 - 오차율임 값이 0이 나왔다면 오차가 크다는 것임

        if aa>100:
            aa= 200 - aa
        else:
            aa=aa
    except:
        print('어깨부위의 사이즈가 없습니다.')
        aa=0
        w1=0
    
    try:
        print(str(data['가슴'].iloc[num] - b)+'cm' '차이납니다')
        chest_num = data['가슴'].iloc[num] 
        
        if chest_num <=chest_intrv0:
            print("매우 타이트")
        elif chest_intrv0 < chest_num <= chest_intrv1:
            print("타이트")
        elif chest_intrv1 < chest_num <= chest_intrv2:
            print("적당")
        elif chest_intrv2 < chest_num <= chest_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        if abs(data['가슴'].iloc[num]-b) > b:
            bb=0
        else:
            bb = 100 - ((data['가슴'].iloc[num]-b)/b)*100

        if bb>100:
            bb= 200 - bb
        else:
            bb=bb
    except:
        print('가슴부위의 사이즈가 없습니다.')
        bb=0
        w2=0
    
    try:
        print(str(data['암홀'].iloc[num] - c)+'cm' '차이납니다')
        armhole_num = data['암홀'].iloc[num] 
        
        if armhole_num <=armhole_intrv0:
            print("매우 타이트")
        elif armhole_intrv0 < armhole_num <= armhole_intrv1:
            print("타이트")
        elif armhole_intrv1 < armhole_num <= armhole_intrv2:
            print("적당")
        elif armhole_intrv2 < armhole_num <= armhole_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['암홀'].iloc[num]-c) > c:
            cc=0
        else:
            cc = 100 - ((data['암홀'].iloc[num]-c)/c)*100
        
        if cc>100:
            cc= 200 - cc
        else:
            cc = cc
    except:
        print('암홀의 사이즈가 없습니다.')
        cc=0
        w3=0
    
    #팔길이는 멘트를 주지 않고 cm차이만 주고 이후 전체 fit percentage를 구할 때는 percentage가
    #필요하므로 그에 대한 식만 구한 뒤 멘트가 아닌 ~cm 차이를 보여준다.
    try:
        print(str(data['허리'].iloc[num] - d)+'cm' '차이납니다')
        waist_num = data['허리'].iloc[num] 
        
        if waist_num <=waist_intrv0:
            print("매우 타이트")
        elif waist_intrv0 < waist_num <= waist_intrv1:
            print("타이트")
        elif waist_intrv1 < waist_num <= waist_intrv2:
            print("적당")
        elif waist_intrv2 < waist_num <= waist_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['허리'].iloc[num]-d) > d:
            dd=0
        else:
            dd = 100 - ((data['허리'].iloc[num]-d)/d)*100
        
        if dd>100:
            dd= 200 - dd
        else:
            dd = dd
    except:
        print('암홀의 사이즈가 없습니다.')
        dd=0
        w4=0
        
    try:
        print(str(data['엉덩이'].iloc[num] - e)+'cm' '차이납니다')
        hip_num = data['엉덩이'].iloc[num] 
        
        if hip_num <=hip_intrv0:
            print("매우 타이트")
        elif hip_intrv0 < hip_num <= hip_intrv1:
            print("타이트")
        elif hip_intrv1 < hip_num <= hip_intrv2:
            print("적당")
        elif hip_intrv2 < hip_num <= hip_intrv3:
            print("살짝 큼")
        else:  
            print("매우 큼")
            
        
        if abs(data['엉덩이'].iloc[num]-e) > e:
            ee=0
        else:
            ee = 100 - ((data['엉덩이'].iloc[num]-e)/e)*100
        
        if ee>100:
            ee= 200 - ee
        else:
            ee = ee
    except:
        print('암홀의 사이즈가 없습니다.')
        ee=0
        w5=0
        
    try:
        if abs(data['바지총장'].iloc[num]-f) > f:
            ff=0
        else:
            ff = 100 - ((data['바지총장'].iloc[num]-f)/f)*100
        print('아우터의 총장은 당신의 신체와 '+ str(round(data['바지총장'].iloc[num]-f,1)) + 'cm'+' 차이납니다.' )

        if ff > 100:
            ff = 200 - ff
        else:
            ff = ff
    except:
        print('총장의 사이즈가 없습니다.')
        ff = 0
        w6 = 0
    
    #너무 오차가 많이 나게 된다면 음수값이 나올수가 있기 때문에
    mean = (aa*w1 + bb*w2 + cc*w3 + dd*w4 + ee*w5 + ff*w6)/(w1+w2+w3+w4+w5+w6)
    print('back-end에 들어가는 실제 percentage = ' + str(round(mean,1))+' %')
    
    if 0<= mean < 20:
        return '당신의 신체사이즈와 60% 부합합니다'
    elif 20 <= mean <40:
        return '당신의 신체사이즈와 65 % 부합합니다'
    elif 40 <= mean <60:
        return '당신의 신체사이즈와 70 % 부합합니다'
    elif 60 <= mean <80:
        return '당신의 신체사이즈와 75 % 부합합니다'
    else:
        return '당신의 신체사이즈와' + str(round(mean,1)) + '% 부합합니다.' 
    
#퍼센티지로 안하니깐 모순이 생김.. 옷이 신체보다 클때랑 작을때랑 비교가 안됨
#차이를 둬야할 것 같은데 퍼센티지가 작다고해서 슬림핏이 아님 마찬가지로 크다고 해서 박시한 핏이 아님

#어차피 백단에 들어가는 내용이겠지만
#신체사이즈 퍼센티지를 80%까지는 percentage그대로 보여주고 80%보다 아래의percentage를 받았다면
#그때는 조건문을 넣어서 60%~80%를 작위적으로 받도록하자
#이유는 오버핏을 제외하고는 신체와 그렇게 차이나는 경우가 없어 사이즈가 있는 모든 옷들은 왠만해서는
#80%이상의 신체 percentage를 가진다 하지만 오버핏인 경우에만 80%아래로 내려가는 경우 혹은
#말도안되게 옷과 신체치수가 차이나는 경우에 80%아래로 나는 경우가 존재하므로 이와 같은 방법을
#이용하면 좋을 듯
```

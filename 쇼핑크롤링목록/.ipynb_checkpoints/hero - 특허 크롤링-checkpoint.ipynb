```python
patent = pd.DataFrame({'app_date':[],
                      'app_name':[],
                      'enroll':[],
                      'summary':[]})
while True:
    page_len = len(browser.find_elements_by_xpath('//*[@class = "board_pager02"]/a'))
    
    for i in range(2,page_len+1):
        browser.find_elements_by_xpath('//*[@class = "board_pager02"]/a[' + str(i)+']')[0].click()
        time.sleep(1.5)
        
        #페이지당 특허 갯수
        page_t = len(browser.find_elements_by_xpath('//*[@class="snb_scroll"]/ul/li'))
        for pt in range(1,page_t+1):
            #각 특허에 들어가는 것
            browser.find_elements_by_xpath('//*[@class="snb_scroll"]/ul/li[' + str(pt) + ']')[0].click()
            time.sleep(0.5)
            #frame접속
            browser.switch_to_frame(browser.find_element_by_name('ifrmDetailArea'))
            html0 = browser.page_source
            time.sleep(0.5)
            html1 = BeautifulSoup(html0,'html.parser')
            
            #본문내용 텍스트 리스트
            html2 = html1.find('div',{'class':'detial_plan_info'}).find_all('li')
            
            #출원인 = 3/ 등록번호 =4 / 출원번호 = 2/ 요약
            #출원번호
            chul_date = ''.join(re.findall('[0-9()\.]',html2[2].text))
            #출원인
            chul_name = ''.join(re.findall('[가-힣]',html2[3].find('span').text))
            #등록번호
            go_num = ''.join(re.findall('[0-9()\.]',html2[4].text))
            #요약 text 수집
            summary_t = html1.find('div',{'class':'detial_plan_txt'}).text ; summary_t
            
            #iframe구조 빠져나오기
            browser.switch_to_default_content()
            
            insert_data = pd.DataFrame({'app_date':[chul_date],
                                          'app_name':[chul_name],
                                          'enroll':[go_num],
                                          'summary':[summary_t]})
            patent = patent.append(insert_data)

    if page_len < 5:
        break
    
```

```python
base_url = 'http://www.kipris.or.kr/khome/main.jsp'
browser = Chrome()
browser.maximize_window()
browser.get(base_url)
```

```python
def application(search):
    #특허 실용신안 카테고리 들어가기(메인화면에서)
    patent = pd.DataFrame({'app_date':[],
                      'app_name':[],
                      'enroll':[],
                      'summary':[]})
    browser.find_elements_by_xpath('//*[@id = "gnavigation"]/li[1]/ul/li[1]/a')[0].click()
    #등록만 뽑아내기
    browser.find_elements_by_xpath('//*[@id = "snavigation"]/ul/li[2]/div[1]//fieldset[3]/div/p/span[1]/input')[0].click()
    browser.find_elements_by_xpath('//*[@id = "snavigation"]/ul/li[2]/div[1]//fieldset[3]/div/p/span[8]/input')[0].click()

    #검색어 입력
    browser.find_elements_by_xpath('//*[@id = "divSearchItems"]/form[2]/fieldset/input')[0].send_keys(search)
    browser.find_elements_by_xpath('//*[@id = "divSearchItems"]/form[2]/fieldset/input')[0].send_keys(Keys.ENTER)
    
    time.sleep(2)
    #특허 한 곳 접속
    browser.find_elements_by_xpath('//*[@name = "listForm"]/section/article[1]/div[1]/h1/a[2]')[0].click()
    #새로운 창에 들어가기
    browser.switch_to_window(browser.window_handles[1])

    while True:
        page_len = len(browser.find_elements_by_xpath('//*[@class = "board_pager02"]/a'))
    
        for i in range(2,page_len+1):
            browser.find_elements_by_xpath('//*[@class = "board_pager02"]/a[' + str(i)+']')[0].click()
            time.sleep(1.5)

            #페이지당 특허 갯수
            page_t = len(browser.find_elements_by_xpath('//*[@class="snb_scroll"]/ul/li'))
            for pt in range(1,page_t+1):
                #각 특허에 들어가는 것
                browser.find_elements_by_xpath('//*[@class="snb_scroll"]/ul/li[' + str(pt) + ']')[0].click()
                time.sleep(0.5)
                #frame접속
                browser.switch_to_frame(browser.find_element_by_name('ifrmDetailArea'))
                html0 = browser.page_source
                time.sleep(0.5)
                html1 = BeautifulSoup(html0,'html.parser')

                #본문내용 텍스트 리스트
                html2 = html1.find('div',{'class':'detial_plan_info'}).find_all('li')

                #출원인 = 3/ 등록번호 =4 / 출원번호 = 2/ 요약
                #출원번호
                chul_date = ''.join(re.findall('[0-9()\.]',html2[2].text))
                #출원인
                chul_name = ''.join(re.findall('[가-힣]',html2[3].find('span').text))
                #등록번호
                go_num = ''.join(re.findall('[0-9()\.]',html2[4].text))
                #요약 text 수집
                summary_t = html1.find('div',{'class':'detial_plan_txt'}).text ; summary_t

                #iframe구조 빠져나오기
                browser.switch_to_default_content()

                insert_data = pd.DataFrame({'app_date':[chul_date],
                                              'app_name':[chul_name],
                                              'enroll':[go_num],
                                              'summary':[summary_t]})
                patent = patent.append(insert_data)

        if page_len < 5:
            break
        
    patent.index = range(len(patent))
    browser.switch_to_window(browser.window_handles[0])
    return patent




```

```python
patent_data = application('패션*추천시스템')
```

```python
import requests # 웹 페이지 소스를 얻기 위한 패키지(기본 내장 패키지이다.)
from bs4 import BeautifulSoup # 웹 페이지 소스를 얻기 위한 패키지, 더 간단히 얻을 수 있다는 장점이 있다고 한다.
from datetime import datetime                                # (!pip install beautifulsoup4 으로 다운받을 수 있다.)
import pandas as pd # 데이터를 처리하기 위한 가장 기본적인 패키지
import time # 사이트를 불러올 때, 작업 지연시간을 지정해주기 위한 패키지이다. (사이트가 늦게 켜지면 에러가 발생하기 때문)
import urllib.request #
from selenium.webdriver import Chrome
import json
import re     
```

```python
shop = pd.read_csv('브랜디솔루션_최종.csv',engine = 'python')
```

```python
shop_m_email = pd.DataFrame({'url':[],
                              'name':[],
                              'solution':[],
                              'email':[]})
```

```python
for i in range(len(shop)):
    try:
        url = shop['url'].iloc[i]
        name = shop['name'].iloc[i]
        solution = shop['solution'].iloc[i]
        
        req = requests.get(url)
        
        if req == 200 :
            soup = BeautifulSoup(req.content, 'html.parser')
        else :
            req = requests.get(url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                    'Chrome/71.0.3578.98 Safari/537.36')})
            soup = BeautifulSoup(req.content, 'html.parser')
            
        page_source = req.text
        email0 = re.findall('[a-zA-Z0-9+-_.]+@[a-zA-Z0-9-]+\.[a-zA-Z0-9-.]+',page_source)
        #자칫 .을 하나만쓰면 ~~기때문에 \을 앞에 붙임

        if email0 == []:
            b = 'none_email'
        else:
            email2 = list(set(email0))

            b=''
            for i in range(len(email2)):
                if i == len(email2)-1:
                    a=email2[i]
                else:
                    a = email2[i] + '  '
                b = b+a
                
        insert_data = pd.DataFrame({'url':[url],
                                    'name':[name],
                                    'solution':[solution],
                                    'email':[b]})
        
        shop_m_email = shop_m_email.append(insert_data) 
    except:
        #아예 접속이 안되는 url -> 이것도 예외처리로 봅아내면 될 듯
        url = shop['url'].iloc[i]
        name = shop['name'].iloc[i]
        solution = shop['solution'].iloc[i]
        b = 'request X'
        insert_data = pd.DataFrame({'url':[url],
                                    'name':[name],
                                    'solution':[solution],
                                    'email' : [b]})
        shop_m_email = shop_m_email.append(insert_data)
        
shop_m_email.index = range(len(shop_m_email))    
    
```

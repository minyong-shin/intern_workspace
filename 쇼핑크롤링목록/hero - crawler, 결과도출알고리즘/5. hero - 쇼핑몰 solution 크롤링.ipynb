```python
#네이버 쇼핑몰 예외처리해야함
shop_solution = pd.DataFrame({'url':[],
                             'solution':[]})
for i in range(len(shop)):
    try:
        if 'naver' in shop['쇼핑몰URL'].iloc[i]:
            url = shop['쇼핑몰URL'].iloc[i]
            solution = 'naver'
            insert_data = pd.DataFrame({'url':[url],
                                        'solution':[solution]})
        else:
            url = shop['쇼핑몰URL'].iloc[i]
            req = requests.get(url)
            if req == 200 :
                soup = BeautifulSoup(req.content, 'html.parser')
            else :
                req = requests.get(url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                        'Chrome/71.0.3578.98 Safari/537.36')})
                soup = BeautifulSoup(req.content, 'html.parser')
            page_source = req.text
            match = re.search(r'cafe24',page_source)
                #match = re.search(r'var EC_ROOT_DOMAIN = (.+);',page_source)#보류

            if match:
                solution = 'cafe24'
            else:
                match = re.search(r'makeshop',page_source)
                if match:
                    solution = 'makeshop'
                else:
                    match = re.search(r'godohosting',page_source)
                    if match:
                        solution = 'godomall'
                    else:
                        solution = 'another solution'
                    
            insert_data = pd.DataFrame({'url':[url],
                            'solution':[solution]})
            
        shop_solution = shop_solution.append(insert_data) 
    except:
         #아예 접속이 안되는 url -> 이것도 예외처리로 봅아내면 될 듯
        url = shop['쇼핑몰URL'].iloc[i]
        solution = 'request x'
        insert_data = pd.DataFrame({'url':[url],
                            'solution':[solution]})
        shop_solution = shop_solution.append(insert_data) 
shop_solution.index = range(len(shop_solution))
```

```python
shop_m_solution = pd.DataFrame({'url':[],
                             'solution':[],
                             'name':[]})
```

```python
#네이버 쇼핑몰 예외처리해야함
for i in range(len(shop)): #1755가 문제 / 
    try:
        #네이버 카페에서 운영하고 있는 것들은 제외
        if 'naver' in shop['url'].iloc[i]:  #네이버카페나 블로그에서 운영중인 쇼핑몰은 url에 naver가 존재
            url = shop['url'].iloc[i]
            name = shop['쇼핑몰'].iloc[i]
            solution = 'naver'
            insert_data = pd.DataFrame({'url':[url],
                                        'solution':[solution],
                                        'name':[name]})
        #본격적인 솔루션 탐색
        else:
            url = shop['url'].iloc[i]
            name = shop['쇼핑몰'].iloc[i]
            req = requests.get(url) #url request 호출
            
            if req == 200 : #정상적으로 requests
                soup = BeautifulSoup(req.content, 'html.parser')
            else : #request가 200이 아닌 403 이나 500, 400 같은 request에러가 발생했을 때 
                #user-agent를 통해서 우회접근을 해야한다.
                req = requests.get(url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                        'Chrome/71.0.3578.98 Safari/537.36')})
                soup = BeautifulSoup(req.content, 'html.parser')
                
            page_source = req.text #정상적으로 호출된 pagesource를 문자열로 바꿔준다
            
            #정규표현식으로 cafe24찾기
            match = re.search(r'cafe24',page_source)
            #match = re.search(r'var EC_ROOT_DOMAIN = (.+);',page_source)#보류

            if match:
                solution = 'cafe24'
            else:
                match = re.search(r'makeshop',page_source)
                if match:
                    solution = 'makeshop'
                else:
                    match = re.search(r'godohosting',page_source)
                    if match:
                        solution = 'godomall'
                    else:
                        solution = 'another solution' #위의 3개의 솔루션 어느 것도 아닐 때
                    
            insert_data = pd.DataFrame({'url':[url],
                            'solution':[solution],
                            'name':[name]})
            
        shop_m_solution = shop_m_solution.append(insert_data) 
    except:
         #아예 접속이 안되는 url -> 이것도 예외처리로 봅아내면 될 듯
        url = shop['url'].iloc[i]
        name = shop['쇼핑몰'].iloc[i]
        solution = 'request x'
        insert_data = pd.DataFrame({'url':[url],
                            'solution':[solution],
                            'name':[name]})
        shop_m_solution = shop_m_solution.append(insert_data) 
shop_m_solution.index = range(len(shop_m_solution))
```

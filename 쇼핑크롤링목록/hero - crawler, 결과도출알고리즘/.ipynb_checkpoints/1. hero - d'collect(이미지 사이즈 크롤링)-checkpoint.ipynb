```python
import requests # 웹 페이지 소스를 얻기 위한 패키지(기본 내장 패키지이다.)
from bs4 import BeautifulSoup # 웹 페이지 소스를 얻기 위한 패키지, 더 간단히 얻을 수 있다는 장점이 있다고 한다.
from datetime import datetime                                # (!pip install beautifulsoup4 으로 다운받을 수 있다.)
import pandas as pd # 데이터를 처리하기 위한 가장 기본적인 패키지
import time # 사이트를 불러올 때, 작업 지연시간을 지정해주기 위한 패키지이다. (사이트가 늦게 켜지면 에러가 발생하기 때문)
import urllib.request #
from selenium.webdriver import Chrome
import json
import re     
```

```python
data_outer = pd.DataFrame({'name':[],
                          'img': [],
                          'url':[],
                          'price':[],
                          'size':[]})

base_url = "http://www.dcollect.co.kr"

req = requests.get(base_url)
if req == 200 :
    soup = BeautifulSoup(req.content, 'html.parser')
else :
    req = requests.get(base_url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup = BeautifulSoup(req.content, 'html.parser')

menu0 = soup.find('nav',{'class':'topmenu'})
menu1 = menu0.find_all('dt')

brand_category = []
brand_name = []
for i in range(len(menu1)):
    brand_category.append(base_url + menu1[i].find('a')['href'])
    brand_name.append('d_collect '+menu1[i].text)
brand_category2 = brand_category[2:13]
base_url2 = brand_category2[0]

req2 = requests.get(base_url2)
if req2 == 200 :
    soup2 = BeautifulSoup(req2.content, 'html.parser')
else :
    req2 = requests.get(base_url2, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup2 = BeautifulSoup(req2.content, 'html.parser')

category_list0 = soup2.find('div',{'class':'xans-element- xans-product xans-product-menupackage '})
category_list2 = category_list0.find_all('li')

category_url = []
for i in range(len(category_list2)):
    category_url.append(base_url + category_list2[i].find('a')['href'])

start_url = category_url[0]
page_url = category_url[0] + '&page=' + '{}'

req3 = requests.get(start_url)
if req3 == 200 :
    soup3 = BeautifulSoup(req3.content, 'html.parser')
else :
    req3 = requests.get(start_url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup3 = BeautifulSoup(req3.content, 'html.parser')

page0 = soup3.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'})
last_page = start_url+page0.find('a',{'class':'last'})['href']


req4 = requests.get(last_page)
if req4 == 200 :
    soup4 = BeautifulSoup(req4.content, 'html.parser')
else :
    req4 = requests.get(last_page, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup4 = BeautifulSoup(req4.content, 'html.parser')

if soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li')[0].text=='1':
    page =  len(soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li'))
elif soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li')[0].text=='6':
    page = 5+len(soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li'))
elif soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li')[0].text=='11':
    page = 10 + len(soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li'))
elif soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li')[0].text=='16':
    page = 15 + len(soup4.find('div',{'class':'xans-element- xans-product xans-product-normalpaging ga09-paging'}).find_all('li'))
else:
    page = 21

for  i in range(page):
    req5 = requests.get(page_url.format(str(i)))
    if req5 == 200 :
        soup5 = BeautifulSoup(req5.content, 'html.parser')
    else :
        req5 = requests.get(page_url.format(str(i)), headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                'Chrome/71.0.3578.98 Safari/537.36')})
        soup5 = BeautifulSoup(req5.content, 'html.parser')

    product0 = soup5.find('ul',{'class':'grid3'})
    product = product0.find_all('li',{'class':'xans-record-'})
    
    for i in range(len(product)):
        try:
            name = product[i].find('div',{'class':'name'}).find('span').text
            
            if 'set' in name or '세트' in name or 'SET' in name:
                pass
            else:
                img = product[i].find('div',{'class':'prdline hoverimg'}).find('img')['src']

                price = product[i].find('span',{'class':'price'}).text

                url = base_url+product[i].find('div',{'class':'prdline hoverimg'}).find('a')['href']

                req6 = requests.get(url)
                if req6 == 200 :
                    soup6 = BeautifulSoup(req6.content, 'html.parser')
                else :
                    req6 = requests.get(url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
                    soup6 = BeautifulSoup(req6.content, 'html.parser')

                size0 = soup6.find('div',{'id':'prddetailimg'}).find_all('img')    

                size1 = []
                for i in range(len(size0)):
                    size1.append(size0[i]['src'])

                matchers = ['INFO'] #단위와 어깨/암홀로 한 차이가 약 30상품(에러가 30개)차이가 남
                size = [s for s in size1 if any(xs in s for xs in matchers)][0]

                insert_data = pd.DataFrame({'name':[name],
                                  'img': [img],
                                  'url':[price],
                                  'price':[url],
                                  'size' : [size]})
                data_outer = data_outer.append(insert_data)
        except:
            pass
data_outer['category'] = 'Outer'
data_outer.index = range(len(data_outer))
    
    
```

```python
img = Image.open('더블 브레스티드 자켓(SV4JK82)size9.jpg')
text = pytesseract.image_to_string(img,lang='kor')

size_table = re.split('\n',text)
matchers = ['사이즈(','어깨너비','가슴둘레','소매길이','총장'] #단위와 어깨/암홀로 한 차이가 약 30상품(에러가 30개)차이가 남
size = [s for s in size_table if any(xs in s for xs in matchers)]
#예외처리해야할 부분이 존재함 -> 사이즈가 리스트로 정리하지 못하게 아예단이 띄어져있는 경우가 있음
#규칙파악해서 조건문으로 처리해야할 듯

size2 = []
for i in range(len(size)):
    size2.append(size[i].split())
    
if '총장' and '소매길이' and '어깨너비' and '가슴둘레' not in size2[1][0] + size2[2][0]+size2[3][0]+size2[4][0]:
    pass
else:
    match = re.match(r"([가-힣]+)([(0-9].+)", size2[0][0], re.I)
    if match:
        items = match.groups()
    size2[0][0] = items[0]

    size3 = list(map(list, zip(*size2))) #리스트를 transpose하는 법
    #import numpy as np
    #np.array(a).T.tolist() - 똑같은 방법  -> array 변환 -> 전치 -> tolist
    df_size = pd.DataFrame(size3[1:], columns = size3[0])
```

```python
#하나의 상품당 약 10~15초 소요
outer_size = pd.DataFrame({'name':[],
                          '사이즈':[],
                          '어깨너비':[],
                          '가슴둘레':[],
                          '소매길이':[],
                          '총장':[]})
for i in range(len(name0)):
    try: 
        name = name0[i]
        img = Image.open(name0[i])
        text = pytesseract.image_to_string(img,lang='kor+eng')

        size_table = re.split('\n',text)
        matchers = ['사이즈(','어깨너비','가슴둘레','소매길이','총장'] #단위와 어깨/암홀로 한 차이가 약 30상품(에러가 30개)차이가 남
        size = [s for s in size_table if any(xs in s for xs in matchers)]
        #예외처리해야할 부분이 존재함 -> 사이즈가 리스트로 정리하지 못하게 아예단이 띄어져있는 경우가 있음
        #규칙파악해서 조건문으로 처리해야할 듯

        size2 = []
        for i in range(len(size)):
            size2.append(size[i].split())
        
        #이미지를 인식하는 과정에서 상품마다 같은 구조의 사이즈 이미지표를 가지고 있으나 다르게 인식해서
        #오류가 발생하게 됨 따라서 그것을 처리하는 처리문
        #opencv를 적용시켜서 인식해보면??
        if '총장' and '소매길이' and '어깨너비' and '가슴둘레' not in size2[1][0] + size2[2][0]+size2[3][0]+size2[4][0]:
            pass
        else:
            match = re.match(r"([가-힣]+)([(0-9].+)", size2[0][0], re.I)
            if match:
                items = match.groups()
            size2[0][0] = items[0]

            size3 = list(map(list, zip(*size2))) #리스트를 transpose하는 법
            #import numpy as np
            
            for j in range(len(size3)):
                if j==0:
                    size3[j].append('name')
                else:
                    size3[j].append(name)
            #np.array(a).T.tolist() - 똑같은 방법  -> array 변환 -> 전치 -> tolist
            df_size = pd.DataFrame(size3[1:], columns = size3[0])

            outer_size =  outer_size.append(df_size)
    except:
        pass
outer_size.index = range(len(outer_size))
```

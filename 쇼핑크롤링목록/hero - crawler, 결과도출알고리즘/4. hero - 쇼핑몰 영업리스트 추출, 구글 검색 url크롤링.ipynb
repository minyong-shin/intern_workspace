# 브랜디 쇼핑몰 url 추출

```python
ep_shop = shop_info['쇼핑몰'].tolist()
```

```python
base_url = 'https://www.google.com'
browser = Chrome()
browser.maximize_window()
browser.get(base_url)
```

```python
shop_info2_remain = pd.DataFrame({'shop_name':[],
                          'link':[]})
for i in range(len(ep_shop)): #쇼핑몰 이름만큼 반복
    try:
        shop_name = ep_shop[i] #행별로 쇼핑몰 이름 저장
        
        #검색창 클릭
        browser.find_elements_by_xpath('//*[@class = "tsf nj"]/div[2]/div/div[2]/div/div[1]/input')[0].click()
        #검색기록 지우기
        browser.find_elements_by_xpath('//*[@class = "tsf nj"]/div[2]/div/div[2]/div/div[1]/input')[0].clear()
        time.sleep(0.5)
        
        #검색창에 쇼핑몰이름 입력
        browser.find_elements_by_xpath('//*[@class = "tsf nj"]/div[2]/div/div[2]/div/div[1]/input')[0].send_keys(ep_shop[i])
        #enter
        browser.find_elements_by_xpath('//*[@class = "tsf nj"]/div[2]/div/div[2]/div/div[1]/input')[0].send_keys(Keys.RETURN)
        time.sleep(0.5)
        
        #구글 검색 page source추출
        shop_sc0 = browser.page_source
        shop_sc = BeautifulSoup(shop_sc0,'html.parser')
        #웹검색결과 / 사이트 링크가 있는 웹 검색결과의 공통 요소 추출
        shop_sc2 = shop_sc.find_all('div',{'class':'g'})
        
        #예외처리- 제대로 수집되지 못한 url이 있었기 때문에 remain추출
        try:
            link = shop_sc2[0].find('link')['href'] #사이트 링크가 있는 웹 검색결과는 바로 뽑아짐
        except:
            link = ''#단순 웹검색결과일 때는 error가 뜨므로 error이 반환되었을 때 link에 빈 문자열을 넣어줌
        try:
            if link == '': #빈문자열일 때는 웹 검색결과 이므로 다른 방법으로 수집해야함
                #다른 방법
                shop_sc2 = shop_sc.find('div',{'class':'rc'})
                link = shop_sc2.find('div',{'class':'r'}).find('a')['href']
        except:
            pass
        
        insert_data = pd.DataFrame({'shop_name':[shop_name],
                          'link':[link]})
        shop_info2_remain = shop_info2_remain.append(insert_data)
    except:
        pass
shop_info2_remain.index = range(len(shop_info2_remain))
```

```python
data_outer = pd.DataFrame({'name':[],
                          'img':[],
                          'price':[],
                          'url':[]})

for i in range(0,5):
    req3 = requests.get(page_url.format(str(i)))
    if req3 == 200 :
        soup3 = BeautifulSoup(req3.content, 'html.parser')
    else :
        req3 = requests.get(page_url.format(str(i)), headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                'Chrome/71.0.3578.98 Safari/537.36')})
        soup3 = BeautifulSoup(req3.content, 'html.parser')
        
    product = soup3.find_all('div',{'class':'product_item gtm_prod'})
    img_list = soup3.find_all('img',{'onerror':'imgError(this);'})
    img_list2 = []
    for i in range(1,len(img_list)+1,2):
        try:
            img_list2.append(img_list[i]['src'])
        except:
            img_list2.append(img_list[i]['data-original'])

    for i in range(len(product)):
        i2=str(i)
        product_url = soup3.find('a',{'onclick':"fnGTMProductClickEvent({})".format("'" + i2 + "'")})['href']

        price = product[i]['data-price']

        name = product[i]['data-name']

        img = img_list2[i]

        insert_data = pd.DataFrame({'name':[name],
                              'img':[img],
                              'price':[price],
                              'url':[product_url]})
        data_outer = data_outer.append(insert_data)
        
data_outer.index = range(len(data_outer))    
```

```python
outer_size = pd.DataFrame({'기준':[],
                          '가슴둘레':[],
                          '허리둘레':[],
                          '총길이':[]})

for i in range(50):
    try:
        samp = data_outer['url'].iloc[i]
        name = data_outer['name'].iloc[i]
        req4 = requests.get(samp)
        if req4 == 200 :
            soup4 = BeautifulSoup(req4.content, 'html.parser')
        else :
            req4 = requests.get(samp, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                    'Chrome/71.0.3578.98 Safari/537.36')})
            soup4 = BeautifulSoup(req4.content, 'html.parser')
        table0 = soup4.find('table',{'class':'tablearea typewidth'})
        table1=table0.find('tbody').find_all('tr')

        size3=[]
        for j in range(len(table1)):
            size0 = table1[j].find_all(['th','span'])
            size2=[]
            for i in range(len(size0)):
                size2.append(size0[i].text)
            size3.append(size2)

        size = list(map(list, zip(*size3))) #리스트를 transpose하는 법
        aa= pd.DataFrame(size[1:],columns=size[0])
        aa['name'] = name
        outer_size = outer_size.append(aa)
    except:
        pass
outer_size.index = range(len(outer_size))
```

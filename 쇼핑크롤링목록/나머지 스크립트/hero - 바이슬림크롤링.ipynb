```python
base_url = 'http://www.byslim.com'

data = pd.DataFrame({"date" : [],
                    "image" : [],
                    "name" : [],
                    "price" : [],
                    "link" : []})

req = requests.get(base_url)
if req == 200 :
    soup = BeautifulSoup(req.content, 'html.parser')
else :
    req = requests.get(base_url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup = BeautifulSoup(req.content, 'html.parser')

table = soup.find_all('li',{'class':'-depth xans-record'})

for category_name in table :
    cn = category_name.text  # 해당 텍스트가 img 태그의 alt 에 있을 수도 있고, 그냥 텍스트로 존재할 수 있다.
    if 'new' in cn.lower() :
        new_category = category_name

new_category.find('a')['href']
new_url = base_url + new_category.find('a')['href']

new_res=requests.get(new_url)

if new_res==200:
    soup2 = BeautifulSoup(new_res.content,'html.parser')
else:
    new_res2 = requests.get(new_url,headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup2 = BeautifulSoup(new_res2.content,'html.parser')

page1 = soup2.find('div',{'class':'xans-element- xans-product xans-product-normalpaging -pagenation'})
page2 = page1.find_all('li',{'class','xans-record-'})
page2[0].find('a')['href']

for i in range(1,len(page2)+1):
    page_url = new_url+ '?page='+'[]'
    
    np = requests.get(page_url.format(str(i)))
    if np ==200:
        soup3 = BeautifulSoup(np.content,'html.parser')
    else:
        np2 = requests.get(page_url.format(str(i)),headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
        soup3 = BeautifulSoup(np2.content,'html.parser')
    
    products = soup3.find_all('div',{'class':'-item xans-record-'})

    for i in products:
        try:
            img = i.find('div',{'class':'-thumb'})
            imgf = img.find('img')['src']

            price_0 = i.find('div', {'rel':'판매가','class' : 'xans-record-'}) # 가격정보가 있는 곳의 클래스 # 원 가격, 할인가격 등 다양한 가격이 있는데, 할인가격으로 표시해주는 것
            price_2 = price_0.find_all('span')[2].text

            name = i.find('div', {'class' : '-name'}).find_all('span')[2].text
            name = re.sub('[^가-힣0-9a-zA-Z\\s\+\[\]\,\.\=\-\_\~\!\@\#\$\%\^\&\*\(\)\;\:\?]','',name)

            #해당 태그에서 text만 가져와라# 유니코드로 변환하기 어려운 특수문자가 있는 경우 글자가 깨진다. 따라서, 필요한 글자들만 정규표현식으로 표시해준다.

            link = i.find('div',{'class','-name'}).find('a')['href']
            link_url = base_url + link

            insert_data = pd.DataFrame({"date" : datetime.today().strftime("%Y-%m-%d"),
                                "image" : [imgf],
                                "name" : [name],
                                "price" : [price_2],
                                "link" : [link_url]})
            data = data.append(insert_data)

        except:
            pass
    
data = data.drop_duplicates(subset='image', keep='last') # 중복된 부분이 있다면 가장 최근것으로 대체
data.index = range(len(data))

try :
    pred_data = pd.read_csv('byslim.csv', encoding='cp949') # 이전에 있던 파일을 포함해 조정하는 것
    all_data = pred_data.append(data)
    data = all_data.drop_duplicates(subset='image', keep='last')
    data.index = range(len(data))
except :
    pass


```

```python
prds = []
index = []
for i in range(len(data)) :
    prd = data['link'][i]
    ind = data.index[i]
    if prd not in prds : # 중복되는 상품이 없도록 url을 수집한다.
        prds.append(prd) 
        index.append(ind) # 해당 상품에 대한 인덱스 정보가 있어야 이후에 옷 이미지와 매칭시킬 수 있으므로 인덱스 정보까지 포함한다.

browser = Chrome() # 크롬 브라우저를 연다.

for web in range(len(prds)) : # 저장된 url에서 사이즈표를 추출하기 위해 반복문을 사용한다.
        try:
            num = index[web]
            browser.get(prds[web]) # prds에 저장된 url을 순서대로 접속한다.
            time.sleep(5) # 작업 지연시간을 지정해주기 위한 것.(없으면 브라우저를 열기도 전에 class를 탐색해 에러가 발생할 것이다.)

            dsize = browser.find_elements_by_css_selector('div.dsize_dsize') # 사이즈표가 하나의 html안에 있는 것, 바로 접근가능
            iframe = browser.find_elements_by_css_selector('#sf_chart_iframe_id') # 사이즈표가 또다른 html안에 있는 것, 우회해서 접근해야한다.

            browser.switch_to_frame(browser.find_element_by_css_selector('#sf_chart_iframe_id')) 
                # 또 다른 html에 접근하기 위해 해당 html에 대한 정보를 가지고 있는 iframe id 를 획득
                # .switch_to_frame 은 iframe 내부에 접근하는 것.
            source = browser.page_source # iframe 내부로 접근하면 텍스트 형식이기 때문에, 이를 html로 파싱해주어야 한다.

            html = BeautifulSoup(source, 'html.parser')

            tbody = html.find('tbody') # 사이즈표가 있는 tbody 태그를 추출
            thead = html.find('thead')

            th = thead.find_all('th')
            td = tbody.find_all('td') # 사이즈표의 행을 의미하는 tr을 추출
            div = len(td) # txt가 하나의 리스트로 저장이 되기 때문에 
                # 이를 데이터프레임화해주기 위해 리스트를 잘라서 다시 저장하여야 한다. 따라서, 사이즈표의 열개수를 지정해주는 것
            txt0 = []
            for i in range(len(th)):
                for j in range(len(th[i])):
                    tx0 = th[i].text
                    txt0.append(tx0)

            txt = [0]
            for i in range(len(td)) :
                for j in range(len(td[i])) :
                    tx = td[i].text # td에 저장되어 있는 텍스트 데이터를 가지고 와서 txt에 저장한다.
                    txt.append(tx)

            txt[0] = txt0


            df_iframe = pd.DataFrame( columns = txt[0])
            df = df_iframe.append(pd.Series(txt[1:], index=txt[0]), ignore_index=True)
            
            
            #왜 밑의 코드가 실행이 안되는 거지...
            #df_iframe = pd.DataFrame(data[1:], columns = data[0]) # Multi list를 DataFrame으로 바꿔주는 것

            name = 'prod_' + ' outer_' + data["name"].iloc[web] +str(num) +'_'+ str(web) + '_size' + '.csv' 
                # ex) prod_top_index_prds_size.csv 형식으로 저장한다. num 은 각각의 상품에 대한 인덱스를 의미, web은 각각의 다른 상품을 의미
            df.to_csv(name, mode='w', encoding='ANSI',index=False) # 데이터프레임을 csv형식으로 저장
        except:
            pass
```

```python
prds = []
index = []
names = []
for i in range(len(data_outer)) :
    prd = data_outer['link'][i]
    ind = data_outer.index[i]
    name = data_outer['name'][i]
    if prd not in prds : # 중복되는 상품이 없도록 url을 수집한다.
        prds.append(prd) 
        index.append(ind) # 해당 상품에 대한 인덱스 정보가 있어야 이후에 옷 이미지와 매칭시킬 수 있으므로 인덱스 정보까지 포함한다.
        names.append(name)
        
df_outer = pd.DataFrame({'사이즈':[],
                '어깨너비':[],
                '가슴둘레':[],
                '밑단둘레':[],
                '소매길이':[],
                '화장':[],
                '총길이':[]})
for web in range(len(prds)):
    try:
        test_item = prds[web]

        req6 = requests.get(test_item)
        if req6 == 200 :
            soup6 = BeautifulSoup(req6.content, 'html.parser')
        else :
            req6 = requests.get(test_item, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                                        'Chrome/71.0.3578.98 Safari/537.36')})
            soup6 = BeautifulSoup(req6.content, 'html.parser')
        size0 = soup6.find('div',{'class':'section size_detail pt0'})
        size1 = size0.find('table',{'class':'hor'})
        
        th0 = size1.find('thead')
        th1 = th0.find_all('th')
        th = []
        for i in range(len(th1)):
            th.append(th1[i].text)
            
        tbody = size1.find('tbody') 
        tr0 = tbody.find_all('tr')
        
        tr1 = []
        for i in range(len(tr0)):
            tr1.append(tr0[i].find_all('td'))
        
        tr2 = []
        for i in range(len(tr1)):
            for j in range(len(tr1[i])):
                tr2.append(tr1[i][j].text)
                
        size_num  = []
        for j in range(len(tr2)):
            bb = re.findall('\d.+',tr2[j])
            if bb==[]:
                bb=['0']
            else:
                pass
            size_num.append(bb[0])
            
        data = [th]
        start_pos=0
        for idx in range(start_pos, len(size_num), len(th)): # 리스트를 잘라서 다시 저장하는 것
                output = size_num[start_pos:start_pos + len(th)]
                if output : #해당 조건문은 어떤걸 뜻하는 것이지...
                    data.append(output)
                start_pos = start_pos + len(th)
                

        df_iframe = pd.DataFrame( columns = data[0])
        for i in range(1,len(data)):
            df = df_iframe.append(pd.Series(data[i], index=data[0]), ignore_index=True)
            df["name"] = names[web]
            df_outer = df_outer.append(df)
        
    except:
        pass
```

```python
req2 = requests.get('https://www.zara.com/kr/ko/woman-outerwear-l1184.html')
if req2 == 200 :
    soup2 = BeautifulSoup(req2.content, 'html.parser')
else :
    req2 = requests.get('https://www.zara.com/kr/ko/woman-outerwear-l1184.html', headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup2 = BeautifulSoup(req2.content, 'html.parser')
```

```python
soup2
```

```python
product0 = soup2.find('div',{'data-controller':'catalog/products-category-controller'})
```

```python
product1 = product0.find('section',{'class':'_customHeight '})
```

```python
product2 = product1.find('ul',{'class':'product-list _productList '})
```

```python
product2
```

```python
product3 = product2.find_all('li',{'class':['product _product ','product _product _more-colors ']})
```

```python
product3
```

```python
data = pd.DataFrame({"date" : [],
                    "image" : [],
                    "name" : [],
                    "price" : [],
                    "link" : []})
 
for i in range(len(product3)):
    
    name = product3[i].find('div',{'class':'product-info-item product-info-item-name'}).text
    
    price = product3[i].find('div',{'class':'price _product-price'}).find('span')['data-price']
    
    url = product3[i].find('div',{'class':'product-info-item product-info-item-name'}).find('a')['href']
    
    req5 = requests.get(url)
    if req5 == 200 :
        soup5 = BeautifulSoup(req5.content, 'html.parser')
    else :
        req5 = requests.get(url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
        soup5 = BeautifulSoup(req5.content, 'html.parser')

    image = soup5.find('div',{'class':'media-wrap image-wrap'}).find('a')['href']
    #사이트를 일정이상 못 긁게 막아놈 - 셀레늄 써야함
    
    insert_data = pd.DataFrame({"date" : datetime.today().strftime("%Y-%m-%d"),
                                    "image" : [image],
                                    "name" : [name],
                                    "price" : [price],
                                    "link" : [url]})
        
    data = data.append(insert_data)


data.index = range(len(data))
```

```python
def replaceMultiple(mainString, toBeReplaces, newString):
    # Iterate over the strings to be replaced
    for elem in toBeReplaces :
        # Check if string is in the main string
        if elem in mainString :
            # Replace the string
            mainString = mainString.replace(elem, newString)
    
    return  mainString
```

```python
req = requests.get(start_url)
if req == 200 :
    soup = BeautifulSoup(req.content, 'html.parser')
else :
    req = requests.get(start_url, headers = {'user-agent': ('Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko)'
                                                                            'Chrome/71.0.3578.98 Safari/537.36')})
    soup = BeautifulSoup(req.content, 'html.parser')
    
category_html = soup.find('ul', {'id' : 'navHeader'})

high_name = ['WOMEN','MEN','KIDS','BABY','COMPANY']
category_list = category_html.find_all('div', {'class' : ['gnb_2016_col col3', 'gnb_2016_col col4', 'gnb_2016_col col5', 'gnb_2016_col col6']})
category_hname = np.repeat(high_name, [4,4,4,2,4], axis=0)

ct_a = []
category_name = []
category_link = []
for category_tab in range(len(category_list[:14])) :
    ul = category_list[category_tab].find_all('ul')
    n = category_hname[category_tab]
    for h in ul :
        a = h.find_all('a')
        for b in a :
            if 'UT' in b.text :
                pass
            else :
                link = b['href']
                ct_link = base_url + link
                category_link.append(ct_link)
                ct_name = n + '_' + b.text
                ct_name = replaceMultiple(ct_name, ['/',"\\"], '_')
                category_name.append(ct_name)
                
category_data = pd.DataFrame({'category_name' : category_name, 'category_link' : category_link})

products_link = []
products_name = []
for length in range(len(category_link)) :
    category_name = category_data['category_name'][length]
    category_req = requests.get(category_data['category_link'][length])
    products_html = BeautifulSoup(category_req.content, 'html.parser')
    
    products0 = products_html.find_all('p', {'class' : 'tumb_img'})
    
    for p in products0 :
        products1 = p.find('a')['href']
        products2 = base_url + products1
        products_link.append(products2)
        products_name1 = p.find('img')['alt']
        products_name1 = replaceMultiple(products_name1, ['/',"\\"], '_')
        products_name.append(category_name + '_' + products_name1)
    category_req.close()
products = pd.DataFrame({'products_name' : products_name, 'products_link' : products_link})
```

```python
size_db = pd.DataFrame({"name":[],
                    "size" : [],
                    "어깨너비" : [],
                    "가슴너비" : [],
                    "소매길이" : [],
                    "전체길이B" : []})

for leng in range(len(products_outer)) :
    products_name = products_outer['products_name'][leng]
    browser2.get(products_outer['products_link'][leng])
    size = browser2.find_element_by_css_selector('div.linkMore')#selectSizeDetail > div.linkMore
    size_link = size.find_element_by_tag_name('a')

    size_link.send_keys(Keys.ENTER)
    browser2.implicitly_wait(delay)

    browser2.switch_to.window(browser2.window_handles[1])#after clicking the link

    size_source = browser2.page_source
    size_html = BeautifulSoup(size_source, 'html.parser')

    size_table = size_html.find('div', {'class' :'size_table'})
    table = size_table.find_all('tr')

    tds = []
    for i in table :
        td = i.find_all(['th', 'td'])
        tds.append(td)

    for i in range(len(tds)):
        for j in range(len(tds[i])):
            try:
                pit = int(tds[i][j]['colspan'])
                pit2 = pit-1
                tds[i] = tds[i]+[tds[i][j].text]*pit2
            except:
                pass


    tx = []
    txt = []

    for i in range(len(table)) :
        for j in range(len(tds[i])) :
            if len(tds[i][j])>=2:
                tx=tds[i][j]
                txt.append(tx)
            else:
                tx = tds[i][j].text # td에 저장되어 있는 텍스트 데이터를 가지고 와서 txt에 저장한다.
                txt.append(tx)
    txt[0]='size'
    start_pos = 0

    end_pos = len(txt)

    div = len(table[0].find_all('th')) # txt가 하나의 리스트로 저장이 되기 때문에 
    # 이를 데이터프레임화해주기 위해 리스트를 잘라서 다시 저장하여야 한다. 따라서, 사이즈표의 열개수를 지정해주는 것

    data = []

    for idx in range(start_pos, end_pos + div, div): # 리스트를 잘라서 다시 저장하는 것
        output = txt[start_pos:start_pos + div]
        if output :
            data.append(output)
        start_pos = start_pos + div

    df_size = pd.DataFrame(data[1:], columns = data[0])

    #왜 이부분 넣으면 크롤링 안되지..
    df_size2 = df_size.set_index('size').T #여기서 에러가 왜 뜨지
    df_size2['size']=df_size2.index
    df_size2.index = range(len(df_size2))
    df_size2['name'] = products_name

    size_db = size_db.append(df_size2)


    browser2.switch_to.window(browser2.window_handles[0])#before clicking the link


size_db.index = range(len(size_db))
```

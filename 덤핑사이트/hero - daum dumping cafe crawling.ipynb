```python
bw_data = pd.DataFrame({'post_title':[],
                       'post_nick':[],
                       'post_date':[],
                       'post_url':[]})
for i in range(1,6):
    browser.find_elements_by_xpath('//*[@class="paging "]/a[' + str(i) +']')[0].click()
    time.sleep(2)
    html0 =  browser.page_source
    html_bw = BeautifulSoup(html0,'html.parser')
    
    post_list0 = html_bw.find('table',{'class':'bbsList'})
    post_list = post_list0.find_all('tr',{'class':True})
    
    insert_data = pd.DataFrame({'post_title':[],
                       'post_nick':[],
                       'post_date':[],
                       'post_url':[]})
    for j in range(len(post_list)):
        post_title = post_list[j].find('td',{'class':'subject'}).find('a').text
        post_nick = post_list[j].find('td',{'class':'nick'}).find('a').text
        post_date = post_list[j].find('td',{'class':'date'}).text
        post_url = base_url0 + post_list[j].find('td',{'class':'subject'}).find('a')['href']
        
        insert_data2 = pd.DataFrame({'post_title':[post_title],
                       'post_nick':[post_nick],
                       'post_date':[post_date],
                       'post_url':[post_url]})
        insert_data = insert_data.append(insert_data2)
        
    bw_data = bw_data.append(insert_data)
    
bw_data.index = range(len(bw_data))
        
        
```
